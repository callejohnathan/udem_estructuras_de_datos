## Punto 1 — Filtrar valores mayores que K (Lista Enlazada Simple)

Dada una **lista enlazada simple de números enteros**, elimine **todos los nodos cuyo valor sea mayor que un número `K`**.

La función debe modificar la lista **sin crear una nueva lista**.

## Entrada

Una lista enlazada simple:

```
4 → 7 → 2 → 9 → 1 → 6
```

y un número:

```
K = 5
```

## Salida esperada

```
4 → 2 → 1
```





## Punto 2 — Detectar el primer valor repetido (Lista Enlazada Simple)

Dada una **lista enlazada simple de números enteros**, determine **cuál es el primer valor que aparece más de una vez en la lista**.

Si ningún valor se repite, la función debe retornar `None`.

## Entrada

Lista enlazada:

```
3 → 7 → 4 → 9 → 7 → 2
```

## Salida

```
7
```

porque es el **primer valor que aparece nuevamente** al recorrer la lista.

## Otro ejemplo

Entrada:

```
1 → 5 → 8 → 3
```

Salida:

```
None
```


## Punto 3 — Verificar si la lista es palíndroma (Lista Doblemente Enlazada)

Dada una **lista doblemente enlazada**, determine si la secuencia de valores es **palíndroma**, es decir, si se lee igual **de izquierda a derecha y de derecha a izquierda**.

La solución debe aprovechar la naturaleza **doblemente enlazada** de la estructura.

## Entrada

Lista doble:

```
1 ⇄ 4 ⇄ 7 ⇄ 4 ⇄ 1
```

## Salida

```
True
```

## Otro ejemplo

Entrada:

```
2 ⇄ 3 ⇄ 5 ⇄ 6
```

Salida:

```
False
```
