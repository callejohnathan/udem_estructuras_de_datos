## **Ejercicio Preparatorio – Recursión con Matrices y Strings (non-tail)**


**Enunciado:**
Se tiene una **matriz NxM** de caracteres (strings de longitud 1).
Dada una posición inicial `(i, j)`, se desea obtener dos cadenas:

1. **Cadena horizontal izquierda:** formada desde la posición `(i, j)` hasta la celda más a la izquierda posible de esa misma fila.
2. **Cadena vertical hacia abajo:** formada desde la posición `(i, j)` hasta la última celda hacia abajo en esa misma columna.

La solución **debe implementarse con recursión non–tail (o "sin cola")**.

---

### **Ejemplo**

**Entrada:**

```python
matriz = [
    ['C', 'A', 'S', 'A'],
    ['L', 'O', 'S', 'O'],
    ['P', 'E', 'R', 'O']
]
i, j = 1, 2  # Letra 'S'
```

**Salida esperada:**

```
Horizontal izquierda: "S O L"
Vertical abajo: "S R"
```

*(sin espacios en el código, los he puesto solo para lectura)*

---

### **Restricciones**

1. La función recibirá: `matriz`, `i`, `j` y debe retornar una **tupla** `(cadena_horizontal, cadena_vertical)`.
2. El recorrido horizontal siempre va de derecha a izquierda dentro de la misma fila hasta la primera columna.
3. El recorrido vertical siempre va hacia abajo dentro de la misma columna hasta la última fila.
4. **No usar ciclos** (`for`, `while`).
5. Implementar cada recorrido como una función recursiva **no–tail**.

---




````python
from typing import Tuple

def e1(matrix: list[list[str]], i:int, j:int) -> tuple[str, str]:
  if(i == len(matrix)):
    return "","" #esto es importante para respetar el retorno de la función y el "pedacito" que necesita left y down.

  if(j == -1):
    return "",""

  left = matrix[i][j] + e1(matrix, i, j-1)[0] #aquí sólo me interesa lo primero que retorna la tupla
  down = matrix[i][j] + e1(matrix, i+1, j)[1] #aquí sólo me importa la parte derecha

  return left, down #recordar que python empaqueta retornos múltiples como tupla --> es lo mismo que poner (left, down)

matriz = [
    ['C', 'A', 'S', 'A'],
    ['L', 'O', 'S', 'O'],
    ['P', 'E', 'R', 'O']
]
i, j = 1, 2  # Letra 'S'
e1(matriz, i, j)
````

## **Ejercicio Preparatorio – Recursión de Cola (Tail Recursion)**


**Enunciado:**
Dado un número entero positivo `n` y un dígito `k` (0 ≤ k ≤ 9), se desea contar **cuántos dígitos de `n` son mayores que `k`** usando **recursión de cola**.

> En este ejercicio, la llamada recursiva debe ser **la última operación** que realiza la función.

---

### **Ejemplo**

**Entrada:**

```
n = 582491
k = 4
```

**Salida esperada:**

```
3
```

*(porque los dígitos mayores que 4 son: 5, 8 y 9)*

---

### **Restricciones**

1. **Prohibido** usar conversión directa a string para recorrer los dígitos.
2. Debe ser **tail recursion**.


````python
def e2(n:int, k: int, cont: int = 0) -> int:
  if(n == 0): #incluye el caso donde n es cero o k es cero.
    return cont

  current_digit: int = n % 10
  if(current_digit > k): cont += 1

  return e2(n // 10, k, cont) #clave recortar el número para llegar al caso base

e2(10031, 1)
````


## **Ejercicio Preparatorio – Manejo de Strings y Posiciones**


**Enunciado:**
Dado un string `s` y un número entero positivo `n`, se desea construir un **nuevo string** formado únicamente por los caracteres de `s` que se encuentren en posiciones **múltiplos de `n`**.
Las posiciones se cuentan desde `0`.

La solución debe implementarse de forma que:

* No se utilicen ciclos (`for`, `while`).
* Se resuelva utilizando **recursión** (puede ser tail o non-tail según se indique).
* El resultado sea un string nuevo que contenga solo los caracteres filtrados.

---

### **Ejemplo**

**Entrada:**

```
s = "programacion"
n = 3
```

**Salida esperada:**

```
"pgmi"
```

*(porque los caracteres en posiciones 0, 3, 6 y 9 son: 'p', 'g', 'm', 'i')*

---

### **Restricciones**

1. No usar funciones integradas como `enumerate`, `filter` o slicing avanzado.

---


````python
#versión tail-recursion
def e3_tail(string: str, n: int, idx: int = 0, new_string: str = "") -> str:
  if(idx == len(string)):
    return new_string

  if(idx % n == 0):
    new_string += string[idx] #si estoy parado en un caracter válido, lo añado a la respuesta

  return e3_tail(string, n, idx+1, new_string)


s = "ay muchachos..."
n = 3
e3_tail(s, n)
````

````python
#versión non-tail-recursion
def e3_nontail(string: str, n: int, idx: int = 0) -> str:
  if(idx == len(string)):
    return ""

  if(idx % n == 0):
    return string[idx] + e3_nontail(string, n, idx+1) #si estoy parado en un caracter válido, lo añado a la respuesta (que se terminará de formar cuando me devuelva)
  return e3_nontail(string, n, idx+1)


s = "ay muchachos..."
n = 4
e3_nontail(s, n)
````
