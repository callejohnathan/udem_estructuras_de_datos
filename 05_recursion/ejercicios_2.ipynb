### 4️⃣ **Contar vocales en un string**
Implementa una función recursiva que cuente el número total de vocales en un string.

**Ejemplo de entrada y salida:**
```python
contar_vocales("recursion")  # Output: 4
contar_vocales("python")     # Output: 1
```


````python
def contar_vocales(string: str, idx: int = 0, conteo = {"a":0, "e":0, "i":0, "o":0, "u":0}) -> dict:
    if(idx == len(string)):
        return conteo
    if(string[idx] in conteo):
        conteo[string[idx]] += 1
    return contar_vocales(string, idx + 1, conteo)

print(contar_vocales("hola"))
````
