### **1️⃣1️⃣ Búsqueda Recursiva de un Número en una Lista de Listas 🔍**  
**Objetivo:** Implementar un algoritmo **recursivo** que, dada una **lista anidada** (una lista que contiene otras listas dentro, de manera arbitraria), **busque un número `X`** y determine si está presente.  

📌 **Tareas:**  
- Se recibe una **lista anidada** donde los elementos pueden ser **números o más listas anidadas**.  
- Se debe **buscar recursivamente** el valor `X` en **todas las sublistas**.  
- Si el número `X` **se encuentra**, devuelve `True`.  
- Si el número `X` **no está presente**, devuelve `False`.  

📌 **Ejemplo 1:** (Caso con el número presente)  
```plaintext
Entrada:  
lista = [1, [2, [3, 4], 5], [6, [7, [8, 9]]]]  
X = 4  
Salida: True  (El número 4 está dentro de la lista)
```

📌 **Ejemplo 2:** (Caso sin el número)  
```plaintext
Entrada:  
lista = [1, [2, [3, 4], 5], [6, [7, [8, 9]]]]  
X = 10  
Salida: False  (El número 10 no está en la lista)
```

🔹 **Pistas:**  
✅ **Caso base:** Si la lista está vacía, devuelve `False`.  
✅ **Si el primer elemento es un número**, compáralo con `X`.  
✅ **Si el primer elemento es otra lista**, realiza una **llamada recursiva** sobre esa sublista.  
✅ **Repite el proceso para el resto de la lista** hasta encontrar el número o agotar todas las posibilidades.

📌 **Tareas:**  
- Se recibe una **matriz de tamaño `m x n`** y un **número `n`**.  
- Se debe contar cuántas veces aparece `n` en cada **columna** y devolver la **columna con más ocurrencias**.  
- **Restricciones:**  
  ✅ **Solo se puede usar recursión** (sin bucles explícitos).  
  ✅ No se puede usar el operador `in`.  

📌 **Ejemplo 1:** (Caso con múltiples ocurrencias)  
```plaintext
Entrada:  
Matriz =  
[
  [1, 2, 3, 4],  
  [5, 6, 3, 7],  
  [8, 9, 3, 2],

n = 3  

Salida: Columna 3 (3 aparece 3 veces en la tercera columna)

n = 4  

Salida: Columna 2 (4 aparece 3 veces en la segunda columna)
```

🔹 **Pistas:**  
✅ Define una función recursiva que explore una columna a la vez.  
✅ Usa **llamadas recursivas** para recorrer los elementos de cada columna.  
✅ Lleva un contador de ocurrencias por columna y encuentra la columna con el valor máximo.

````python
#Tail - Recursion
from typing import List
x = [2,3,4]
def sumatoria_lista(lista: List[int], i: int, sumatoria: int) -> int:
  if(i == len(lista)):
    return sumatoria

  return sumatoria_lista(lista, i+1, sumatoria + lista[i])

sumatoria_lista(x, 0, 0)
````

````python
#Non-Tail Recursion
from typing import List
x = [2,3,4]
def sumatoria_lista(lista: List[int], i: int) -> int:
  if(i == len(lista)):
    return 0

  return lista[i] + sumatoria_lista(lista, i + 1)

sumatoria_lista(x, 0)
````

````python
def recorrer_matriz(matriz, i=0, j=0) -> None:
  if(i == len(matriz)):
    return None

  if(j == len(matriz[i])): #j ya tiene el tamaño de la fila en la que estamos?
    recorrer_matriz(matriz, i+1, 0)
  else:
    print(matriz[i][j])
    recorrer_matriz(matriz, i, j+1)

matriz = [[1,2,3],[5,4,5],[6,7,8]]
recorrer_matriz(matriz)
````

````python
def sumatoria_matriz(matriz, i=0, j=0, sumatoria = 0) -> int:
  if(i == len(matriz)):
    return sumatoria

  if(j == len(matriz[i])): #j ya tiene el tamaño de la fila en la que estamos?
    return sumatoria_matriz(matriz, i+1, 0, sumatoria)
  else: #estoy en posición válida
    sumatoria += matriz[i][j]
    return sumatoria_matriz(matriz, i, j+1, sumatoria)

matriz = [[1,2,3],[5,4,5],[6,7,8]]
sumatoria_matriz(matriz)
````

````python
#sin cola
def sumatoria_matriz(matriz, i=0, j=0) -> int:
  if(i == len(matriz)):
    return 0

  if(j == len(matriz[i])): #j ya tiene el tamaño de la fila en la que estamos?
    return 0 + sumatoria_matriz(matriz, i+1, 0)
  else: #estoy en posición válida
    return matriz[i][j] + sumatoria_matriz(matriz, i, j+1)

matriz = [[1,2,3],[5,4,5],[6,7,8]]
sumatoria_matriz(matriz)
````
