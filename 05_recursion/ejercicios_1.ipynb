### 1. **Sumatoria de una lista**
   Implementa una función recursiva que reciba una lista de números y devuelva la suma de todos sus elementos.

   ```python
   def sumatoria(lista):
       pass
   ```

   **Ejemplo:**
   ```python
   sumatoria([1, 2, 3, 4, 5])  # Output: 15
   sumatoria([10, -5, 7])      # Output: 12
   ```




````python
from typing import List
def suma_elementos_lista(lista: List[int], pos: int = 0, suma: int = 0) -> int:
  #caso base
  if(pos == len(lista)):
    return suma
  #caso recursivo
  #extraigo el valor
  valor_actual = lista[pos]
  #lo acumulo
  suma += valor_actual
  return suma_elementos_lista(lista, pos + 1, suma)


suma_elementos_lista([1,4,5,6])
````

## Ejercicio de clase: devolver sumatoria de elementos de una lista siempre y cuando los elementos adyacentes tengan una diferencia de 1 ascendentemente

````python
from typing import List
def sumatoria_lista_ascendente(lista: List[int], i: int = 0, suma: int = 0) -> int:
  if(len(lista) == 0 or i == len(lista) - 1):
    return suma

  if(lista[i] == lista[i+1] - 1):
    suma += lista[i] + lista[i+1]

  return sumatoria_lista_ascendente(lista, i+1, suma)

sumatoria_lista_ascendente([4,6,8,4,7,5,6,9,7,8])
````

### Ejercicio de clase: dado un string alfanumérico S, devuelva el string sin números. Además, si el número (dígito) está antes de una letra, replica esa letra la cantidad de veces del dígito.

````python
def eliminar_y_replicar(cadena: str, i: int = 0, nueva_cadena: str = "") -> str:
  #caso base
  if(i == len(cadena)-1):
    if(cadena[i].isalpha()):
      nueva_cadena += cadena[i]
    return nueva_cadena

  #caso recursivo
  char_actual = cadena[i]
  if(char_actual.isalpha()):
    nueva_cadena += char_actual
  else:
    if(cadena[i+1].isalpha()): #es letra lo que me sigue?
      #replicar
      char_actual = int(char_actual) #lo convierto a entero
      nueva_cadena += (cadena[i+1] * (char_actual-1))
    else:
      #ignoro char_actual
      nueva_cadena += ""

  #sigo avanzando
  return eliminar_y_replicar(cadena, i+1, nueva_cadena)


eliminar_y_replicar("a5b45a")
````
