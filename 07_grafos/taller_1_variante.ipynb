## Matriz de adyacencia

```python
from typing import List, Dict, Any
import pprint
class Graph_AM:
  def __init__(self):
    self.adj_matrix: List[List[int]] = []
    self.nodes: List[Any] = []
    self.size: int = 0

  def add_vertex(self, value: Any) -> None:
    #ya está el nodo?
    if(value in self.nodes):
      return None #está repetido

    self.nodes.append(value) #agregué el valor del nodo
    #ajustar mi matriz
    for row in self.adj_matrix:
      row.append(0)
    self.adj_matrix.append([0] * (self.size + 1))
    self.size += 1

  def add_edge(self, vertex_1: Any, vertex_2: Any, directed: bool = True):
    #si no existe vertex_1, lo creo
    if(vertex_1 not in self.nodes):
      self.add_vertex(vertex_1)

    #si no existe vertex_2, lo creo
    if(vertex_2 not in self.nodes):
      self.add_vertex(vertex_2)

    pos_v1 = self.nodes.index(vertex_1)
    pos_v2 = self.nodes.index(vertex_2)

    if(not directed):
      self.adj_matrix[pos_v2][pos_v1] = 1

    self.adj_matrix[pos_v1][pos_v2] = 1


g = Graph_AM()
g.add_edge("b", "a")
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("x", "y", False) # NO DIRIGIDA
g.add_edge("d", "c", False) # NO DIRIGIDA
g.add_edge("c", "z", False) # NO DIRIGIDA
g.add_edge("m", "n", False) # NO DIRIGIDA
g.add_edge("m", "a", False) # NO DIRIGIDA
g.add_edge("m", "b", False) # NO DIRIGIDA
g.add_edge("m", "c", False) # NO DIRIGIDA
g.add_edge("d", "c", False) # NO DIRIGIDA
g.add_edge("d", "a", False) # NO DIRIGIDA
print(g.nodes)
pprint.pp(g.adj_matrix)
```

## Lista de adyacencia

```python
from typing import List, Dict, Any
import pprint
class Graph:
  def __init__(self):
    self.adj_list: Dict[Any,List[Any]] = {}
    self.size: int = 0

  def add_vertex(self, value: Any) -> None:
    #ya está el nodo?
    if(value in self.adj_list):
      return None #está repetido

    self.adj_list[value] = []
    self.size += 1

  def add_edge(self, vertex_1: Any, vertex_2: Any, directed: bool = True):
    #si no existe vertex_1, lo creo
    if(vertex_1 not in self.adj_list):
      self.add_vertex(vertex_1)

    #si no existe vertex_2, lo creo
    if(vertex_2 not in self.adj_list):
      self.add_vertex(vertex_2)

    if(not directed):
      if(vertex_1 not in self.adj_list[vertex_2]):
        self.adj_list[vertex_2].append(vertex_1)
    if(vertex_2 not in self.adj_list[vertex_1]):
        self.adj_list[vertex_1].append(vertex_2)


g = Graph()
g.add_edge("b", "a")
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("z", "a", False) # NO DIRIGIDA
g.add_edge("z", "b") # NO DIRIGIDA
g.add_edge("z", "y") # NO DIRIGIDA
g.add_edge("z", "x") # NO DIRIGIDA
g.add_edge("z", "p") # NO DIRIGIDA
g.add_edge("p", "a") # NO DIRIGIDA
g.add_edge("p", "b") # NO DIRIGIDA
g.add_edge("y", "x", False) # NO DIRIGIDA
print(g.adj_list)
```

```python
g = Graph()
g.add_edge("b", "a")
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("c", "a", False) # NO DIRIGIDA
g.add_edge("z", "a", False) # NO DIRIGIDA
g.add_edge("z", "b") # NO DIRIGIDA
g.add_edge("z", "y") # NO DIRIGIDA
g.add_edge("z", "x") # NO DIRIGIDA
g.add_edge("z", "p") # NO DIRIGIDA
g.add_edge("p", "a") # NO DIRIGIDA
g.add_edge("p", "b") # NO DIRIGIDA
g.add_edge("y", "x", False) # NO DIRIGIDA

def can_reach(graph: Graph, vertex_1: Any, vertex_2: Any) -> bool:
  if(vertex_1 not in graph.adj_list or vertex_2 not in graph.adj_list):
    return False
  visited = set()
  queue = [vertex_1]
  while queue:
    vertex = queue.pop(0)
    if(vertex == vertex_2):
      return True
    for neighbor in graph.adj_list[vertex]:
      if neighbor not in visited:
        visited.add(neighbor)
        queue.append(neighbor)
  return False

print(g.adj_list)
print(can_reach(g, "a", "b"))
```

## Ejercicio 2

* Dado un grafo no dirigido, devolver cuántos componentes desconectados hay. Un componente desconectado es un grupo de nodos (cantidad de nodos >= 1) que no se conecta a otro grupo de nodos dentro del grafo.

----> Dar ejemplo en el tablero

```python
def count_components(graph: Graph, visited: set) -> int:
  def dfs(graph: Graph, vertex: Any, visited: set) -> None:
    visited.add(vertex)
    for neighbor in graph.adj_list[vertex]:
      if neighbor not in visited:
        dfs(graph, neighbor, visited)

  count = 0
  for vertex in graph.adj_list:
    if vertex not in visited:
      dfs(graph, vertex, visited)
      count += 1
  return count

g = Graph()
g.add_edge("ay","muchachos")
g.add_edge("no","cancelemos")
g.add_edge("esto", "se")
g.add_edge("se", "compone")
print(g.adj_list)
print(count_components(g, set()))
```
