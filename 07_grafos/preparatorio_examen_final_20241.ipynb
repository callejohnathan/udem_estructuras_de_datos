Esto significa que cualquier nodo que sea "referido" o "alcanzado" por otros nodos exactamente `n` veces debe ser eliminado del grafo.

#### Parámetros

La función `eliminar_nodos_con_entradas` debe recibir:
1. Una **lista de adyacencia** en forma de diccionario, donde cada clave es un nodo y su valor es una lista de nodos a los que está conectado.

### Ejercicio: 👨‍👩‍👧‍👦 Encontrando a los parientes alcanzables

Dado un grafo ponderado representado mediante una **lista de adyacencia**, cada nodo tiene una **etiqueta** de uno o dos caracteres. Si un nodo tiene una etiqueta de **dos caracteres**, es considerado un "pariente" de otro nodo. El primer carácter de la etiqueta representa el nodo en sí mismo, y el segundo carácter en minúscula indica el **nodo madre** del que es pariente.

Por ejemplo, en un grafo con los nodos **A**, **Xa**, **Za**, y **Wa**:
- **Xa**, **Za**, y **Wa** son parientes de **A**.

### Objetivo de la Función

Implementa una función llamada `encontrar_parientes_alcanzables` que reciba:
1. Una **lista de adyacencia ponderada** en forma de diccionario, donde cada clave es un nodo y su valor es una lista de tuplas (cada tupla contiene un nodo destino y la distancia hacia ese nodo).
2. Un **nodo madre**.

La función debe devolver una lista de **todos los parientes alcanzables** desde el nodo madre. Para cada pariente, debe incluir:
- La etiqueta del pariente.
- La **distancia total** desde el nodo madre hasta el pariente.
- **Todos los caminos posibles** si existen múltiples rutas con diferentes distancias.

### Parámetros de la función

- **grafo**: Un diccionario de lista de adyacencia ponderada, donde cada clave representa un nodo y su valor es una lista de tuplas `(nodo_destino, distancia)`.
- **nodo_madre**: El nodo madre cuya red de parientes queremos explorar.

### Ejemplo de Entrada y Salida 📥📤

1. **Entrada**:
   ```python
   grafo = {
       "A": [("Xa", 3), ("B", 5)],
       "Xa": [("Za", 2)],
       "B": [("Wa", 4)],
       "Za": [],
       "Wa": []
   }
   nodo_madre = "A"
   ```

2. **Salida esperada**:
   ```python
   {
       "Xa": [(3, ["A", "Xa"])],
       "Za": [(5, ["A", "Xa", "Za"])],
       "Wa": [(9, ["A", "B", "Wa"])]
   }
   ```
   
   Explicación:
   - **Xa** es alcanzable desde **A** a una distancia de `3`.
   - **Za** es alcanzable desde **A** a través de **Xa**, a una distancia total de `5`.
   - **Wa** es alcanzable desde **A** a través de **B**, con una distancia total de `9`.
