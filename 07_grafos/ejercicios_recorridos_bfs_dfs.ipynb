# 🧠 Paso a Paso para implementar BFS (Breadth-First Search)

## 📋 Objetivo

Visitar todos los nodos de un grafo **nivel por nivel**, comenzando desde un nodo origen.

## 🛤️ Algoritmo de pensamiento

1. **Crear una estructura para marcar visitados**.
   - Puede ser un `set` o una `lista`.

2. **Crear una cola** para almacenar los nodos por visitar.
   - Al inicio, agrega el nodo de partida.

3. **Mientras la cola no esté vacía:**
   - Extrae el primer nodo de la cola.
   - Marca el nodo como visitado.
   - Procesa el nodo (por ejemplo, imprímelo o guárdalo).
   - Agrega a la cola todos los vecinos **no visitados**.

## 📌 Pensamiento clave

- **Primero en entrar, primero en salir.** (Cola)
- Procesa el nivel completo antes de bajar de nivel.

# 🧠 Paso a Paso para implementar DFS (Depth-First Search)

## 📋 Objetivo

Visitar todos los nodos de un grafo **siguiendo un camino lo más profundo posible**, antes de retroceder.

## 🛤️ Algoritmo de pensamiento

### Opción 1: DFS usando Recursión

1. **Crear una estructura para marcar visitados**.

2. **Definir una función recursiva** que reciba el nodo actual.

3. En cada llamada:
   - Marca el nodo como visitado.
   - Procesa el nodo (por ejemplo, imprímelo o guárdalo).
   - Para cada vecino **no visitado**, llama recursivamente la función.

### Opción 2: DFS usando Pila (sin recursión)

1. **Crear una estructura para marcar visitados**.

2. **Crear una pila**.
   - Al inicio, agrega el nodo de partida.

3. **Mientras la pila no esté vacía:**
   - Extrae el último nodo agregado.
   - Si no está visitado:
     - Márcalo como visitado.
     - Procesa el nodo.
     - Agrega todos sus vecinos a la pila.

## 📌 Pensamiento clave

- **Último en entrar, primero en salir.** (Pila)
- Sigue un camino profundo antes de retroceder.

# 🧩 Consejos para ambas implementaciones

- Usa una estructura para registrar qué nodos ya visitaste y evitar ciclos infinitos.
- Asegúrate de **procesar** un nodo **sólo cuando lo visites por primera vez**.
- En grafos no conexos (grafos donde no todo está conectado), necesitarás recorrer desde **todos los nodos no visitados**.

## 🌟 Introducción

En esta sección exploraremos dos enfoques principales para encontrar caminos mínimos en un grafo:

- **BFS especial** para grafos **no ponderados**
- **Dijkstra** para grafos **ponderados**

Estos algoritmos son esenciales en redes, mapas, planificación de rutas y muchas otras áreas. 🚀

> 📌 **Modelo mental:** Buscar caminos mínimos es como encontrar la ruta más rápida para llegar a tu destino en una ciudad, dependiendo de si todas las calles cuestan lo mismo (BFS) o si algunas son más "caras" o "lentas" (Dijkstra).

---

# 🧠 Distancias mínimas en grafos no ponderados (BFS especial)

## 📋 Idea principal

Cuando **todas las aristas** tienen el **mismo peso** (o no hay pesos), podemos usar **BFS** para encontrar el número mínimo de saltos o movimientos entre nodos.


## 🛤️ Algoritmo de pensamiento

1. Crear una estructura para marcar distancias:
   - Inicializa todas las distancias en infinito (`inf`) o `None`.

2. Crear una **cola** e insertar el nodo de partida con distancia 0.

3. Mientras la cola no esté vacía:
   - Extrae un nodo.
   - Para cada vecino no visitado:
     - Asigna distancia = distancia del nodo actual + 1.
     - Agrega el vecino a la cola.


## 🧠 Claves para recordarlo

- BFS **siempre expande primero los caminos más cortos**.
- No necesitas comparar distancias ni actualizar valores más de una vez.


## 📌 Ejemplo de aplicación

- Encontrar la **cantidad mínima de movimientos** en un tablero.
- Determinar el **nivel de separación** en una red social.

## 🧩 Modelo de Pensamiento para Dijkstra Básico

### 🔍 Objetivo

Encontrar la **distancia mínima** desde un nodo origen a todos los demás nodos en un grafo con pesos **no negativos**.

---

## 🧠 Paso a paso mental

### 1. **Representación del grafo**

Primero, define el grafo usando una estructura que te permita acceder rápidamente a los vecinos y sus pesos. La más común:

```python
graph = {
    "A": [("B", 4), ("C", 2)],
    "B": [("C", 5), ("D", 10)],
    "C": [("D", 3)],
    "D": []
}
```

➡️ Usa un **diccionario de listas de tuplas** para representar el grafo dirigido con pesos.

---

### 2. **Inicialización de estructuras**

Debes seguir tres cosas esenciales:

* Un diccionario `distancia` con valores infinitos (∞) al principio excepto el origen (0).
* Un conjunto o lista `visitados` para saber qué nodos ya se procesaron.
* Una **cola de prioridad** (min-heap) para siempre tomar el nodo con menor distancia actual.

```python
import heapq

distancia = {nodo: float('inf') for nodo in graph}
distancia[origen] = 0

cola = [(0, origen)]  # (distancia, nodo)
```

---

### 3. **Exploración**

Mientras haya nodos por procesar en la cola:

* Saca el nodo con la menor distancia.
* Si ya fue visitado, ignóralo.
* Para cada vecino del nodo:

  * Calcula la **distancia tentativa**.
  * Si es menor que la conocida, **actualiza** y añade a la cola.


---

### 4. **Resultado**

El diccionario `distancia` tendrá el costo mínimo desde el origen a cada nodo.

---

| ------------- | ------------------------------------------------ |
| `distancia`   | Guarda el costo mínimo actual a cada nodo        |

| `visitados`   | Evita ciclos y reevaluaciones inútiles           |
| `graph`       | Representa las conexiones y pesos del grafo      |

## 🧠 Modelo de Pensamiento: Algoritmo de Dijkstra para Ruta más Corta entre V1 y V2

### 🎯 Objetivo

Encontrar la ruta más corta desde un nodo origen `A` hasta un nodo destino `B` minimizando el costo total del camino (suma de los pesos).

---

### 📌 1. **Inicializa las estructuras de control**

* Crea un **diccionario de distancias** para guardar la distancia mínima conocida desde `A` hacia cada nodo.

  ```python
  dist = {nodo: infinito para cada nodo}
  dist[A] = 0
  ```
* Crea una **cola de prioridad (min heap)** que priorice el nodo con menor distancia conocida.
* Crea un **diccionario de predecesores** para reconstruir el camino final.

---

### 📌 2. **Explora los nodos**

Mientras haya nodos en la cola:

#### a) Extrae el nodo con menor distancia acumulada

```python
nodo_actual = heapq.heappop(cola)
```

#### b) Si es el destino `B`, **termina** (ya tienes la ruta más corta).

#### c) Si no, revisa todos sus vecinos:

Para cada vecino `v` de `nodo_actual`:

* Calcula la distancia tentativa:

  ```python
  nueva_distancia = dist[nodo_actual] + peso(nodo_actual, v)
  ```
* Si esta distancia es **menor** que la conocida:

  * Actualiza `dist[v] = nueva_distancia`
  * Actualiza el predecesor de `v`
  * Agrega `v` a la cola con su nueva distancia

---

### 📌 3. **Reconstruye la ruta**

Cuando termines, **reconstruye la ruta** desde `B` hasta `A` usando el diccionario de predecesores:

```python
ruta = []
nodo = B
while nodo:
    ruta.append(nodo)
    nodo = predecesor[nodo]
ruta.reverse()
```

---

### 🛠️ Herramientas necesarias

* Diccionario de distancias `dist`
* Diccionario de predecesores `prev`
* Cola de prioridad `heapq`
* Grafo representado como lista de adyacencia con pesos:

  ```python
  grafo = {
      'A': [('B', 2), ('C', 5)],
      'B': [('C', 1)],
      ...
  }
  ```

---

### 🧩 Consideraciones

* Dijkstra **no funciona** con pesos negativos.
* Es más eficiente si se usa un heap (cola de prioridad).
* Si solo quieres la distancia mínima (no la ruta), puedes omitir `prev`.
