## Ejercicio 1

Una empresa de domicilios quiere saber qué clientes puede atender desde una bodega central sin superar cierto costo máximo de envío. La ciudad se modela como un grafo ponderado G, donde cada nodo representa una ubicación y cada arista representa una ruta entre dos ubicaciones con un costo asociado: distancia, tiempo estimado o valor logístico.

Dado un nodo inicial N, que representa la bodega, y un entero K, que representa el costo máximo permitido, se debe determinar cuáles ubicaciones pueden alcanzarse desde N con un costo total menor o igual a K.

Tu tarea consiste en implementar un algoritmo que, dado un grafo ponderado G, un entero K y un nodo inicial N, devuelva todos los nodos alcanzables desde N con costo acumulado ≤K, sin considerar caminos que generen ciclos. Se puede asumir que todos los pesos del grafo son positivos.

## Ejercicio 2

* Tienes un grafo no dirigido y ponderado. Ten en cuenta:
Cada nodo del grafo contiene una letra (por ejemplo, A, B, C…).
Cada arista entre dos nodos tiene un costo entero positivo.
Además, tienes una lista de palabras escritas en mayúsculas.
Tu tarea será verificar si todas las palabras pueden formarse recorriendo el grafo.

---
### Reglas para formar una palabra

Para formar una palabra, por ejemplo "CAB", debe existir un camino en el grafo que visite nodos con esas letras en ese orden.
  * Ejemplo: un nodo con "C", luego un nodo con "A", luego un nodo con "B".
  * El costo de formar la palabra es la suma de los costos de las aristas que se recorren.
  * Las palabras son independientes entre sí (se pueden reutilizar nodos y aristas en cada palabra).



#### Objetivo

Debes escribir un programa que:
* Revise si todas las palabras se pueden formar en el grafo.
* Si es así, sume los costos de cada palabra y devuelva esa suma total.
* Si alguna palabra no puede formarse, el programa debe devolver -1.
