```python
#representación de clase

#MATRIZ
class GraphM:
    def __init__(self):
        self.adj_matrix = [] #relaciones
        self.nodes = [] #etiquetas

    def add_vertex(self, label):
        if label not in self.nodes:
            self.nodes.append(label)

            #ajustar la matriz
            for row in self.adj_matrix:
                row.append(0)
            self.adj_matrix.append([0] * len(self.nodes))

    def add_edge(self, vertex_1, vertex_2, directed = True):
        if(vertex_1 not in self.nodes):
            self.add_vertex(vertex_1)

        if(vertex_2 not in self.nodes):
            self.add_vertex(vertex_2)

        #estoy completamente seguro que vertex1 y vertex2
        pos_v1 = self.nodes.index(vertex_1)
        pos_v2 = self.nodes.index(vertex_2)

        #ya tengo las dos posiciones->agrego la relación
        self.adj_matrix[pos_v1][pos_v2] = 1
        if(not directed):
            self.adj_matrix[pos_v2][pos_v1] = 1

    def __repr__(self):
        result = "   " + " ".join(self.nodes) + "\n"

        for i, row in enumerate(self.adj_matrix):
            result += self.nodes[i] + "  " + " ".join(map(str, row)) + "\n"

        return result
```

```python
from typing import Any, Dict, List
from collections import deque

class GraphL:
  def __init__(self):
    self.adj_list: Dict[Any, List[Any]] = {}
    self.size: int = 0

  def add_vertex(self, value: Any) -> None:
    if value in self.adj_list:
      return None  # ya está el nodo
    self.adj_list[value] = []
    self.size += 1

  def add_edge(self, vertex_1: Any, vertex_2: Any, directed: bool = True):
    if vertex_1 not in self.adj_list:
      self.add_vertex(vertex_1)
    if vertex_2 not in self.adj_list:
      self.add_vertex(vertex_2)

    # agregar la arista
    if vertex_2 not in self.adj_list[vertex_1]:
      self.adj_list[vertex_1].append(vertex_2)

    # si no es dirigido, agregar también la inversa
    if not directed and vertex_1 not in self.adj_list[vertex_2]:
      self.adj_list[vertex_2].append(vertex_1)

  def __repr__(self):
    result = ""
    for vertex, neighbors in self.adj_list.items():
      result += f"{vertex}: {neighbors}\n"
    return result
```

```python
from typing import Any, Dict, List
from collections import deque

class GraphWL:
  def __init__(self):
    self.adj_list: Dict[Any, List[Any]] = {}
    self.size: int = 0

  def add_vertex(self, value: Any) -> None:
    if value in self.adj_list:
      return None  # ya está el nodo
    self.adj_list[value] = []
    self.size += 1

  def add_edge(self, vertex_1: Any, vertex_2: Any, directed: bool = True, weight: Any = None):
    if vertex_1 not in self.adj_list:
      self.add_vertex(vertex_1)
    if vertex_2 not in self.adj_list:
      self.add_vertex(vertex_2)

    # agregar la arista
    if vertex_2 not in self.adj_list[vertex_1]:
      self.adj_list[vertex_1].append((vertex_2, weight))

    # si no es dirigido, agregar también la inversa
    if not directed and vertex_1 not in self.adj_list[vertex_2]:
      self.adj_list[vertex_2].append((vertex_1, weight))
      
  def __repr__(self):
    result = ""
    for vertex, neighbors in self.adj_list.items():
      result += f"{vertex}: {neighbors}\n"
    return result
```

Debido a restricciones de batería y tiempo, la empresa necesita determinar si existe una ruta que permita al dron visitar todos los puntos exactamente una vez, sin repetir ubicaciones, iniciando desde cualquier punto de la planta.

```python
def get_first_hamiltonian_path(graph):
    def dfs(graph, current, visited):
        if(current in visited):
            return False
        visited.append(current)
        if(len(visited) == len(graph.adj_list)):
            return True
        
        for neighbor in graph.adj_list[current]:
            if(neighbor not in visited):
                if(dfs(graph, neighbor, visited)):
                    return True
                visited.pop()
        return False
    
    for node in graph.adj_list:
        visited = []
        if(dfs(graph, node, visited)):
            return visited
    return []

graph = GraphL()
graph.add_edge("V", "C")
graph.add_edge("A", "A")
graph.add_edge("B", "V")
graph.add_edge("B", "D")
graph.add_edge("V", "X")
graph.add_edge("B", "A")
graph.add_edge("C", "A")
graph.add_edge("A", "D")
graph.add_edge("D", "X")
print(graph)
get_first_hamiltonian_path(graph)
```

### Ejercicio 3

Una empresa de domicilios quiere saber qué clientes puede atender desde una bodega central sin superar cierto costo máximo de envío. La ciudad se modela como un grafo ponderado G, donde cada nodo representa una ubicación y cada arista representa una ruta entre dos ubicaciones con un costo asociado: distancia, tiempo estimado o valor logístico.

Dado un nodo inicial N, que representa la bodega, y un entero K, que representa el costo máximo permitido, se debe determinar cuáles ubicaciones pueden alcanzarse desde N con un costo total menor o igual a K.

Tu tarea consiste en implementar un algoritmo que, dado un grafo ponderado G, un entero K y un nodo inicial N, devuelva todos los nodos alcanzables desde N con costo acumulado ≤K, sin considerar caminos que generen ciclos. Se puede asumir que todos los pesos del grafo son positivos.

```python
def neighbors_within_range(graph, k, current):
    if current not in graph.adj_list:
        return []
    
    queue = deque([(current, 0)])  # (nodo, costo acumulado)
    visited = set()
    while(queue):
        vertex, cost = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            for neighbor, weight in graph.adj_list[vertex]:
                new_cost = cost + weight
                if new_cost <= k and neighbor not in visited:
                    queue.append((neighbor, new_cost))
    visited.remove(current)  # Excluir el nodo de inicio
    return list(visited)
    

graph = GraphWL()
graph.add_edge("A", "A", True, 0)
graph.add_edge("B", "V", True, 5)
graph.add_edge("B", "D", True, 2)
graph.add_edge("V", "X", True, 3)
graph.add_edge("V", "C", True, 9)
graph.add_edge("B", "A", True, 1)
graph.add_edge("C", "A", True, 4)
graph.add_edge("D", "C", True, 4)

print(graph)
neighbors_within_range(graph, 10, "B")
```
