# Abstract Data Types (ADTs) 🚀

## 1. Introducción 🧠

Un **Abstract Data Type (ADT)** es un modelo matemático para tipos de datos definidos por su comportamiento desde el punto de vista del usuario. No define cómo se implementa el tipo de datos, sino qué operaciones soporta y qué resultados se esperan de esas operaciones.

El concepto de **abstracción** en la ciencia de la computación se refiere a la capacidad de ocultar detalles innecesarios y exponer solo lo relevante para el usuario. En los ADTs, esto significa que la estructura interna de los datos no es visible para el usuario, quien solo interactúa con las operaciones definidas.

---

## 2. Importancia de los ADTs 🔍

Los ADTs juegan un papel fundamental en el diseño y desarrollo de software porque permiten separar la implementación de los datos de su uso. Esto trae múltiples ventajas, entre ellas:

- **🧩 Modularidad**: Se pueden desarrollar y mejorar componentes de forma independiente.
- **🔄 Reutilización**: Un ADT puede utilizarse en diferentes contextos sin necesidad de conocer su implementación interna.
- **🛠️ Mantenimiento**: Al definir claramente qué operaciones se pueden realizar sobre un tipo de datos, es más fácil realizar cambios sin afectar el código que los usa.
- **📌 Claridad**: Al enfocarse en el qué y no en el cómo, los ADTs permiten expresar soluciones de manera más clara y concisa.

---

## 3. Ejemplos comunes de ADTs 📂

Algunos de los ADTs más conocidos y utilizados en computación son:

1. **📋 Lista (List)**: Permite almacenar elementos en un orden específico y acceder a ellos mediante una posición.
2. **📬 Cola (Queue) y Cola de prioridad (Priority Queue)**: La cola sigue un orden de FIFO (First In, First Out), mientras que la cola de prioridad extrae elementos según su prioridad.
3. **📦 Pila (Stack)**: Sigue el orden LIFO (Last In, First Out), utilizado en estructuras como la ejecución de funciones en un programa.
4. **🟢 Conjunto (Set)**: Representa una colección de elementos únicos, útil en operaciones de teoría de conjuntos.
5. **🗂️ Diccionario o Mapa (Dictionary / Map)**: Almacena pares clave-valor, facilitando la búsqueda eficiente de información.
6. **🗺️ Grafo (Graph)**: Representa conexiones entre elementos, útil en problemas de redes, optimización y rutas.

---

## 4. Ejemplos de ADTs en la vida real 🌍

Los ADTs no solo se encuentran en la programación, sino que también pueden verse reflejados en la vida cotidiana:

- **📝 Lista de tareas**: Es un ADT tipo lista, donde cada tarea tiene una posición y se puede agregar, eliminar o modificar.
- **🏦 Cola en un banco**: Representa una estructura de cola, donde los clientes son atendidos en el orden en el que llegaron.
- **🍽️ Pila de platos**: Funciona como una pila, donde el último plato en colocarse es el primero en ser tomado.
- **📖 Agenda telefónica**: Es un diccionario, donde los nombres son claves y los números de teléfono son los valores asociados.
- **🚉 Red de transporte**: Es un grafo, donde las estaciones o puntos de conexión son nodos y las rutas entre ellos son las aristas.

Estos ejemplos ilustran cómo los ADTs modelan situaciones del mundo real, proporcionando una base conceptual para el diseño de software estructurado y eficiente.

---

## 5. Conclusión 🎯

Los **Abstract Data Types** permiten diseñar estructuras de datos sin preocuparse inicialmente por su implementación, facilitando la reutilización, la claridad y la modularidad en el desarrollo de software. Al definir claramente las operaciones permitidas y ocultar los detalles internos, los ADTs proporcionan una manera eficiente de modelar datos y resolver problemas computacionales.

6. **Diccionarios o mapas**: Asociación clave-valor y operaciones comunes.

### 🔹 **Conceptos de Abstracción y Encapsulamiento**
7. **Abstracción en programación**: Ocultar detalles de implementación y exponer solo funcionalidades esenciales.
8. **Programación orientada a objetos (POO)**: Clases, objetos y métodos.
9. **Encapsulamiento y visibilidad**: Control de acceso a los datos y métodos (`public`, `private`, `protected`).

10. **Listas enlazadas**: Diferencia entre listas simples, dobles y circulares.
11. **Pilas y colas**: Principios de `LIFO` (Last In, First Out) y `FIFO` (First In, First Out).
12. **Grafos y árboles**: Representación de datos en forma de nodos y aristas.
