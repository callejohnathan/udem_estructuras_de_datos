````python
from typing import List

class EmptyQueue(Exception):
  ...

#ADT: Abstract Data Types --> Contrato que define una colección
class Queue:
  def __init__(self):
    self.__queue: List[int] = [] #Estructura de Datos implementadora

  def enqueue(self, element: int) -> None: #Agrega después del First
    self.__queue.append(element)
    return None

  def dequeue(self) -> int : #Retorna y elimina el First
    if(len(self.__queue) == 0):
      raise EmptyQueue("Cola vacía... No podés hacer dequeue")

    return self.__queue.pop(0)


  def peek(self) -> int: #Retorna el First
    if(len(self.__queue) == 0):
      raise EmptyQueue("Cola vacía... No podés hacer dequeue")

    return self.__queue[0]

  def __repr__(self) -> str:
    return str(self.__queue)

  def __len__(self) -> int:
    return len(self.__queue)

#ADT: Abstract Data Types --> Contrato que define una colección
class PriorityQueue:
  def __init__(self, priority):
    self.__queue: List[int] = [] #Estructura de Datos implementadora
    self.__priority: str = priority

  def enqueue(self, element: int) -> None: #Agrega después del First
    if(self.__priority == "min"):
      self.__queue.append(element)
      self.__queue.sort()
    elif(self.__priority == "max"):
      self.__queue.append(element)
      self.__queue.sort(reverse = True)
    else:
      raise ValueError("Grave... Este tipo de cola no existe...")

  def dequeue(self) -> int : #Retorna y elimina el First
    if(len(self.__queue) == 0):
      raise EmptyQueue("Cola vacía... No podés hacer dequeue")

    return self.__queue.pop(0)


  def peek(self) -> int: #Retorna el First
    if(len(self.__queue) == 0):
      raise EmptyQueue("Cola vacía... No podés hacer dequeue")

    return self.__queue[0]

  def __repr__(self) -> str:
    return str(self.__queue)

  def __len__(self) -> int:
    return len(self.__queue)


class EmptyStack(Exception):
  ...

#STACK: push, pop, peek --> colección: ed. implementadora para guardar los datos
class Stack:
  #estructura --> atributos
  def __init__(self):
    self.__stack: list[int] = [] #estructura de datos implementadora

  #comportamiento --> métodos
  def push(self, element: int) -> None: #agrega después del tope
    self.__stack.append(element)
    return None

  def pop(self) -> int: #retorna y elimina el tope
    if(len(self.__stack) == 0):
      raise EmptyStack("No puedes hacer pop... La pila está vacía")
    return self.__stack.pop()

  def peek(self) -> int: #retorna el tope
    if(len(self.__stack) == 0):
      raise EmptyStack("No puedes hacer peek... La pila está vacía")
    return self.__stack[-1]

  def add_from_list(self, data: list[int]) -> None:
    for i in data:
      self.push(i)

  def __repr__(self) -> str:
    return f"{self.__stack}"

  def __len__(self) -> int:
    return len(self.__stack)
````

### Enunciado: Simulador de Navegación Web

Imagina que estás desarrollando la lógica básica de un navegador web. Un navegador permite al usuario **navegar entre las páginas visitadas** con las acciones clásicas:

* **Ir hacia atrás** (volver a una página anterior en el historial).
* **Ir hacia adelante** (avanzar a la siguiente página en el historial).

#### Tu tarea

Diseña una **función en Python** que reciba:

1. Un **listado de páginas web** (en el orden en que fueron visitadas).
2. Un **punto de partida**, que corresponde al índice de la página actual dentro de la lista.

La función debe simular, dentro de un **ciclo infinito**, la navegación entre estas páginas según las acciones del usuario:

* Si el usuario elige "atrás", debe moverse a la página inmediatamente anterior (si existe).
* Si elige "adelante", debe moverse a la página inmediatamente siguiente (si existe).
* En cada paso, la función debe mostrar cuál es la página en la que se encuentra actualmente.

#### Consideraciones

* Si no hay más páginas hacia atrás o hacia adelante, la función debe mostrar un mensaje indicando que no es posible realizar la acción.
* El orden de la lista original representa exactamente el orden cronológico en que las páginas fueron visitadas.
* El programa debe mantenerse en ejecución, esperando nuevas instrucciones de navegación (ejemplo: una entrada de usuario como `"back"` o `"forward"`).

#### Ejemplo de Entrada

```python
paginas = ["google.com", "udemy.com", "github.com", "stackoverflow.com"]
punto_inicial = 2
```

#### Ejemplo de Comportamiento esperado

```
Página actual: github.com
> back
Página actual: udemy.com
> back
Página actual: google.com
> back
No hay más páginas hacia atrás.
> forward
Página actual: udemy.com
```


````python
from typing import List

def navegador_web(paginas: List[str], punto_inicial: int) -> None:
    """
    Simula la navegación web usando dos pilas para manejar
    las operaciones 'back' y 'forward'.

    Args:
        paginas (List[str]): listado de páginas visitadas en orden cronológico.
        punto_inicial (int): índice de la página actual.

    Returns:
        None: La función ejecuta un ciclo interactivo indefinido
              que muestra la página actual y espera instrucciones.
    """

    # Validación de punto inicial
    if punto_inicial < 0 or punto_inicial >= len(paginas):
        raise ValueError("El punto inicial no es válido para la lista de páginas.")

    # Inicializamos las dos pilas
    back_stack = Stack()
    forward_stack = Stack()

    # Página actual
    actual: str = paginas[punto_inicial]

    # Llenamos la pila de atrás con las páginas antes del punto inicial
    for i in range(punto_inicial):
        back_stack.push(paginas[i])

    # Llenamos la pila de adelante con las páginas después del punto inicial
    for i in range(len(paginas) - 1, punto_inicial, -1):
        forward_stack.push(paginas[i])

    print(f"Página actual: {actual}")

    # Ciclo infinito de interacción
    while True:
        accion: str = input("> ").strip().lower()

        if accion == "back":
            if len(back_stack) > 0:
                # La página actual pasa a la pila de adelante
                forward_stack.push(actual)
                # La nueva página actual viene de la pila de atrás
                actual = back_stack.pop()
                print(f"Página actual: {actual}")
            else:
                print("No hay más páginas hacia atrás.")

        elif accion == "forward":
            if len(forward_stack) > 0:
                # La página actual pasa a la pila de atrás
                back_stack.push(actual)
                # La nueva página actual viene de la pila de adelante
                actual = forward_stack.pop()
                print(f"Página actual: {actual}")
            else:
                print("No hay más páginas hacia adelante.")

        elif accion == "exit":
            print("Cerrando navegador...")
            break

        else:
            print("Acción no reconocida. Use 'back', 'forward' o 'exit'.")


paginas = ["google.com", "udemy.com", "github.com", "stackoverflow.com"]
navegador_web(paginas, 2)
````

````python
from typing import List, Dict, Tuple
import unicodedata
# --- Redefinición de PriorityQueue para poder agregar este tipo de datos ---> hacer hincapié en que los tipos de datos requieren una intervención en la lógica de ordenamiento
# Cola de prioridad implementada como lista ordenada manualmente
class PacientePriorityQueue:
  def __init__(self) -> None:
    self.__queue: List[Tuple[int, int, int, Dict[str, str]]] = []
    self.__orden = 0  # contador de llegada

  def enqueue(self, paciente: Dict[str, str], puntaje: int) -> None:
    self.__orden += 1
    entrada = (-puntaje, -paciente["edad"], self.__orden, paciente)
    self.__queue.append(entrada)
    self.__queue.sort()  # mantiene orden de prioridad

  def dequeue(self) -> Dict[str, str]:
    if not self.__queue:
        raise EmptyQueue("No hay pacientes en espera.")
    return self.__queue.pop(0)[-1]

  def peek(self) -> Dict[str, str]:
    if not self.__queue:
        raise EmptyQueue("No hay pacientes en espera.")
    return self.__queue[0][-1]

  def peek_n(self, n: int) -> List[Dict[str, str]]:
    return [entrada[-1] for entrada in self.__queue[:n]]

  def __len__(self) -> int:
    return len(self.__queue)

# --- Funciones auxiliares ---

def normalizar_texto(texto: str) -> str:
    """
    Elimina acentos y convierte a minúsculas para hacer comparaciones robustas.
    """
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto.lower())
        if unicodedata.category(c) != 'Mn'
    )

def calcular_puntaje(diagnostico: str, pesos: Dict[str, int]) -> int:
    """
    Calcula el puntaje de urgencia de un diagnóstico según el diccionario de pesos.
    - Cuenta cada problemática solo una vez.
    - Ignora acentos y mayúsculas.
    """
    texto = normalizar_texto(diagnostico)
    puntaje = 0
    for problema, valor in pesos.items():
        if normalizar_texto(problema) in texto:
            puntaje += valor
    return puntaje

# --- Simulación del triaje ---

def triaje_pacientes(pacientes: List[Dict[str, str]], pesos: Dict[str, int]) -> None:
    """
    Simula un sistema de triaje interactivo usando una cola de prioridad.
    """
    # Inicializamos la cola
    cola = PacientePriorityQueue()

    # Insertamos pacientes iniciales
    for p in pacientes:
        puntaje = calcular_puntaje(p["diagnostico"], pesos)
        p["puntaje"] = puntaje
        cola.enqueue(p, puntaje)

    # Ciclo interactivo
    while True:
        accion = input("> ").strip().lower()

        if accion == "ver":
            try:
                p = cola.peek()
                print(f"Próximo: {p['nombre']} ({p['edad']}) | puntaje={p['puntaje']} | dx=\"{p['diagnostico']}\"")
            except EmptyQueue:
                print("No hay pacientes para mostrar.")

        elif accion == "atender":
            try:
                p = cola.dequeue()
                print(f"Atendido: {p['nombre']} ({p['edad']}) | puntaje={p['puntaje']}")
            except EmptyQueue:
                print("No hay pacientes para atender.")

        elif accion.startswith("agregar"): #para simplificarlo, pueden asumir que siempre se pide todo.
            try:
                # Formato esperado: agregar nombre,edad,diagnóstico
                datos = accion.replace("agregar", "").strip()
                nombre, edad, dx = [x.strip() for x in datos.split(",", 2)]
                nuevo = {"nombre": nombre, "edad": int(edad), "diagnostico": dx}
                puntaje = calcular_puntaje(dx, pesos)
                nuevo["puntaje"] = puntaje
                cola.enqueue(nuevo, puntaje)
                print(f"Paciente agregado: {nuevo['nombre']} ({nuevo['edad']}) | puntaje={nuevo['puntaje']}")
            except Exception:
                print("Formato inválido. Use: agregar nombre,edad,diagnóstico")

        elif accion.startswith("listar"):
            try:
                n = int(accion.split()[1])
                lista = cola.peek_n(n)
                if not lista:
                    print("No hay pacientes en espera.")
                else:
                    for i, p in enumerate(lista, 1):
                        print(f"{i}) {p['nombre']} ({p['edad']}) | puntaje={p['puntaje']} | dx=\"{p['diagnostico']}\"")
            except Exception:
                print("Formato inválido. Use: listar n")

        elif accion == "salir":
            print("Finalizando triaje...")
            break

        else:
            print("Acción no reconocida. Use 'ver', 'atender', 'agregar', 'listar n' o 'salir'.")

pacientes = [
    {"nombre": "Ana", "edad": 72, "diagnostico": "Desmayos frecuentes y dolor en el corazón"},
    {"nombre": "Luis", "edad": 34, "diagnostico": "Dolor de cabeza intenso y mareo"},
    {"nombre": "Marta", "edad": 65, "diagnostico": "Dolor en el corazón"},
    {"nombre": "Diego", "edad": 50, "diagnostico": "Caída con golpe en la cabeza y náuseas"}
]

pesos = {
    "dolor en el corazón": 10,
    "desmayos frecuentes": 15,
    "mareo": 5,
    "dolor de cabeza": 3,
    "náuseas": 2,
    "golpe en la cabeza": 8
}

triaje_pacientes(pacientes, pesos)
````

### 3️⃣ **Eliminar Estrellas de una Cadena (Removing Stars From a String)**  

💡 **Descripción:**  
Se te da una cadena **s** que contiene letras minúsculas y caracteres `*`.  
Cada `*` representa una operación especial en la que:  

- Se elimina la **estrella** (`*`).  
- Se elimina **el carácter más cercano a su izquierda** (eliminación en pareja).  

Debes devolver la **cadena resultante** después de eliminar **todas** las estrellas siguiendo las reglas anteriores.  

🔹 **Notas:**  
- Siempre es posible realizar la operación según las reglas dadas.  
- La cadena resultante será **única** para cada entrada.  

📌 **Ejemplos:**  

```plaintext

Salida: "lecoe"

Explicación:
1. Se elimina `t` (el carácter más cercano a `*`) → "lee*cod*e"
2. Se elimina `e` (el carácter más cercano a `*`) → "lecod*e"
3. Se elimina `d` (el carácter más cercano a `*`) → "lecoe"
```

```plaintext
Entrada: s = "erase*****"
Salida: ""

Explicación:
1. Se eliminan `e, r, a, s, e` con cada `*`, dejando una cadena vacía.
```

🔎 **Restricciones:**  
- `1 ≤ s.length ≤ 10⁵`  
- `s` contiene solo letras minúsculas (`a-z`) y `*`.  
- Siempre hay suficientes caracteres antes de cada `*` para realizar la operación.  

✅ **Objetivo:**  
Utilizar una **estructura eficiente** para manejar las eliminaciones de manera óptima.  

🚀 **Sugerencia:**  
- **Usar una pila** es una excelente opción. Se pueden almacenar las letras en la pila y cuando se encuentra un `*`, se elimina el último carácter añadido.  
- **Evitar manipulaciones innecesarias de la cadena** para mejorar el rendimiento, especialmente con entradas grandes.

````python
from typing import List

def eliminar_estrellas(cadena: str) -> str:
    """
    Elimina estrellas '*' de una cadena siguiendo las reglas:
    - Se borra la estrella.
    - Se borra el carácter más cercano a su izquierda.

    Args:
        cadena (str): cadena de entrada con letras minúsculas y '*'

    Returns:
        str: cadena resultante sin estrellas y tras aplicar las eliminaciones.
    """
    # Usamos la pila definida previamente
    pila = Stack()

    for ch in cadena:
        if ch == "*":
            # Cada * elimina el último caracter agregado
            pila.pop()
        else:
            # Guardamos las letras en la pila
            pila.push(ch)

    # Reconstruimos el string con los elementos de la pila
    # La pila es una lista interna, usamos join para mayor eficiencia
    return "".join(pila._Stack__stack)  # ⚠️ Acceso al atributo privado solo para este caso


# --- Ejemplos de prueba ---
print(eliminar_estrellas("leet**cod*e"))   # "lecoe"
print(eliminar_estrellas("erase*****"))    # ""
````

### **6️⃣ Deshacer/Rehacer en un Editor de Texto**  
Simula la funcionalidad **deshacer (undo) y rehacer (redo)** en un editor de texto. Usa dos pilas: una para almacenar las acciones realizadas y otra para las acciones que se pueden rehacer.  

Ejemplo de operaciones:  
1. `Escribir "Hola"`  
2. `Escribir " Mundo"`  
3. `Deshacer → "Hola"`  
4. `Rehacer → "Hola Mundo"`  

**Objetivo:** Aplicar pilas en un problema realista de software.

````python
from typing import Optional

def editor_texto() -> None:
    """
    Simula un editor de texto con funcionalidades de deshacer (undo) y rehacer (redo).
    Usa dos pilas: undo_stack y redo_stack.
    """
    undo_stack = Stack()
    redo_stack = Stack()

    # El texto actual (estado inicial vacío)
    texto_actual: str = ""
    undo_stack.push(texto_actual)

    while True:
        accion = input("> ").strip().lower()

        if accion.startswith("escribir"):
            # Extraer el texto después de la palabra "escribir"
            _, contenido = accion.split(" ", 1)
            texto_actual += contenido
            undo_stack.push(texto_actual)
            # Vaciar la pila de redo, ya que hay nueva acción
            redo_stack = Stack()
            print(f"Texto actual: \"{texto_actual}\"")

        elif accion == "undo":
            if len(undo_stack) > 1:  # Siempre queda al menos el inicial ""
                # Mover el estado actual a redo_stack
                redo_stack.push(undo_stack.pop())
                texto_actual = undo_stack.peek()
                print(f"Deshacer → \"{texto_actual}\"")
            else:
                print("No hay más acciones para deshacer.")

        elif accion == "redo":
            if len(redo_stack) > 0:
                estado = redo_stack.pop()
                undo_stack.push(estado)
                texto_actual = estado
                print(f"Rehacer → \"{texto_actual}\"")
            else:
                print("No hay acciones para rehacer.")

        elif accion == "ver":
            print(f"Texto actual: \"{texto_actual}\"")

        elif accion == "salir":
            print("Cerrando editor...")
            break

        else:
            print("Acción no reconocida. Use: escribir <texto>, undo, redo, ver, salir.")

editor_texto()
````

````python
from typing import Tuple, Union

def encontrar_suma(cola: Queue, objetivo: int) -> Union[Tuple[int, int], str]:
    """
    Encuentra dos números dentro de una cola que sumen un valor objetivo.
    Usa únicamente colas como estructura de datos, manteniendo el orden original.

    Args:
        cola (Queue): cola con enteros
        objetivo (int): número a alcanzar con la suma de dos elementos

    Returns:
        Union[Tuple[int, int], str]: tupla con los dos números si existen,
                                     o un mensaje indicando que no hay solución.
    """
    aux = Queue()
    solucion = None

    # Recorremos toda la cola original
    while len(cola) > 0:
        x = cola.dequeue()

        # Revisamos con todos los elementos que ya están en aux
        for _ in range(len(aux)):
            y = aux.dequeue()

            if solucion is None and x + y == objetivo:
                solucion = (x, y)

            aux.enqueue(y)  # devolvemos y a aux

        # Guardamos x en aux para que se mantenga el orden
        aux.enqueue(x)

    # Restaurar la cola original (aux → cola) manteniendo el orden
    while len(aux) > 0:
        cola.enqueue(aux.dequeue())

    return solucion if solucion else f"No existen dos números que sumen {objetivo}."


q = Queue()
for num in [1, 3, 7, 2, 8]:
    q.enqueue(num)

print(q)
print(encontrar_suma(q, 10))  # hay varias validas: (8,2) y  (7,3)
print(q)  # La cola se conserva intacta
````

### Enunciado

Se tiene una **pila de colas**, es decir, una colección organizada como una pila (LIFO) en la que cada elemento es, a su vez, una **cola** con varios datos almacenados.

Dado un elemento `e`, se requiere **determinar si este se encuentra en alguna de las colas** dentro de la pila. En caso de estar presente, el programa deberá retornar la **posición emulada** de dicho elemento como si toda la pila de colas se interpretara como una **matriz** (colección de colecciones).

* La posición se representará mediante dos índices:

  * `i`: índice de la pila (qué cola dentro de la pila lo contiene).
  * `j`: índice dentro de la cola (la posición dentro de esa cola).

Si el elemento no se encuentra en ninguna de las colas, el algoritmo deberá devolver `-1`.

---

### Ejemplo de entrada/salida

**Entrada:**

```
Pila de colas = [
   [5, 8, 10], #esto sería un objeto de tipo cola lleno con 5,8,10
   [2, 4, 6], #esto sería un objeto de tipo cola lleno con 2,4,6
   [7, 9] #esto sería un objeto de tipo cola lleno con 7,9
] #esto es una pila con las tres colas mencionadas antes
e = 4
```

**Salida:**

```
(1, 1)
```

````python
from typing import Tuple, Union

def buscar_en_pila_de_colas(pila: Stack, e: int) -> Union[Tuple[int, int], int]:
    """
    Busca un elemento en una pila de colas y devuelve su posición emulada (i, j).
    Si no se encuentra, devuelve -1.

    Args:
        pila (Stack): pila donde cada elemento es una cola (Queue).
        e (int): elemento a buscar.

    Returns:
        Union[Tuple[int, int], int]: posición (i, j) o -1 si no está.
    """
    pila_aux = Stack()  # para restaurar después
    posicion: Union[Tuple[int, int], int] = -1
    i = 0  # índice de la pila (0 es el fondo, como matriz)

    # Pasamos todo a pila_aux para recorrer en orden
    while len(pila) > 0:
        cola: Queue = pila.pop()

        cola_aux = Queue()
        j = 0  # índice dentro de la cola

        encontrado = False

        # Recorremos cada cola
        while len(cola) > 0:
            elem = cola.dequeue()
            if posicion == -1 and elem == e:
                posicion = (i, j)
                encontrado = True
            cola_aux.enqueue(elem)
            j += 1

        # Restaurar la cola original
        while len(cola_aux) > 0:
            cola.enqueue(cola_aux.dequeue())

        # Guardar la cola en la pila auxiliar
        pila_aux.push(cola)

        i += 1

    # Restaurar la pila original
    while len(pila_aux) > 0:
        pila.push(pila_aux.pop())

    return posicion

# Creamos las colas
c1 = Queue()
c1.enqueue(5)
c1.enqueue(8)
c1.enqueue(10)

c2 = Queue()
c2.enqueue(2)
c2.enqueue(4)
c2.enqueue(6)

c3 = Queue()
c3.enqueue(7)
c3.enqueue(9)

# Creamos la pila de colas
pila = Stack()
pila.push(c1)
pila.push(c2)
pila.push(c3)

print("antes",pila)
print(buscar_en_pila_de_colas(pila, 4))  # (1, 1)
print(buscar_en_pila_de_colas(pila, 9))  # (2, 1)
print(buscar_en_pila_de_colas(pila, 11)) # -1
print("después",pila) #se mantiene intacta
````
