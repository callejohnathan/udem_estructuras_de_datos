### 🔥 **¡Reto: La Matriz de Personas!** 🔥  

👥 **Desafío:**  

📌 **Paso 1:**  
🛠️ **Construye una matriz de tamaño NxN** llena de objetos de tipo **Persona**.  
Cada persona debe tener:  
✔️ **Nombre**  
✔️ **Edad**  
✔️ **Un saludo personalizado** con su nombre y edad.  

📌 **Paso 2:**  
🧠 **Analiza la matriz** y haz que **saluden aquellas personas** que sean **mayores que todos sus vecinos adyacentes** (arriba, abajo, izquierda y derecha).  

📌 **Paso 3:**  
📊 **Cuenta cuántas personas cumplen con esta condición** y retorna ese número.  

🎯 **Objetivo:** Implementar funciones para cada uno de los pasos descritos. Ten muy en cuenta el uso de typehints y apóyate en dunder methods que puedan ser útiles. 🚀

```python
import random
from typing import List
#Definir la clase --> para crear los objetos

class Persona:
  def __init__(self, nombre: str, edad: int):
    self.nombre: str = nombre
    self.edad: int = edad

  def saludar(self) -> str:
    return f"Hola, me llamo, {self.nombre}"

  def __repr__(self) -> str:
    return f"{self.edad} - {self.nombre}"

def crear_lista(n: int) -> List[Persona]:
  personas: List[Persona] = []
  for i in range(n):
    #crear objeto
    persona = Persona(f"Persona{i}", random.randint(20,100))
    #agregar objeto
    personas.append(persona)
  return personas


# Terminar!!!
def comparar_para_saludar(personas: List[Persona]) -> None:
  for i in range(len(personas)):
    if(personas[i].edad > personas[i+1].edad and personas[i].edad > personas[i-1].edad):
      personas[i].saludar()
```
