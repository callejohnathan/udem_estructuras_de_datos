## Lista Enlazada Simple

````python
from typing import Optional, Any

class NodeSingly:
  def __init__(self, data: Any, next = None) -> None:
    self.data = data
    self.next = next

  def __repr__(self) -> str:
    return f"{self.data}"

class SinglyLinkedList:
  def __init__(self) -> None:
    self.head: Optional[NodeSingly] = None
    self.tail: Optional[NodeSingly] = None

  def append(self, data: Any) -> None:
    new_node = NodeSingly(data)
    if not self.head:
      self.head = self.tail = new_node
    else:
      assert self.tail is not None
      self.tail.next = new_node
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    current = self.head
    while current:
      values.append(str(current.data))
      current = current.next
    return " → ".join(values) if values else "[]"
````

## Lista Doblemente Enlazada

````python
from typing import Optional, Any

class NodeDoubly:
  def __init__(self, data: Any, prev = None, next = None) -> None:
    self.data = data
    self.prev = prev
    self.next = next

  def __repr__(self) -> str:
    return f"{self.data}"

class DoublyLinkedList:
  def __init__(self) -> None:
    self.head: Optional[NodeDoubly] = None
    self.tail: Optional[NodeDoubly] = None

  def append(self, data: Any) -> None:
    new_node = NodeDoubly(data)
    if not self.head:
      self.head = self.tail = new_node
    else:
      assert self.tail is not None
      self.tail.next = new_node
      new_node.prev = self.tail
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    current = self.head
    while current:
      values.append(str(current.data))
      current = current.next
    return " ⇄ ".join(values) if values else "[]"
````

## Lista Enlazada Circular

````python
from typing import Optional, Any

class NodeCircular:
  def __init__(self, data: Any, next = None) -> None:
    self.data = data
    self.next = next

  def __repr__(self) -> str:
      return f"{self.data}"

class CircularLinkedList:
  def __init__(self) -> None:
    self.head: Optional[NodeCircular] = None
    self.tail: Optional[NodeCircular] = None

  def append(self, data: Any) -> None:
    new_node = NodeCircular(data)
    if not self.head:
      self.head = self.tail = new_node
      new_node.next = new_node  # apunta a sí mismo
    else:
      assert self.tail is not None
      new_node.next = self.head
      self.tail.next = new_node
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    if not self.head:
      return "[]"
    current = self.head
    while True:
      values.append(str(current.data))
      current = current.next
      if current == self.head:
        break
    return " → ".join(values) + " → (head)"
````

## Punto 1 — Filtrar valores mayores que K (Lista Enlazada Simple)

Dada una **lista enlazada simple de números enteros**, elimine **todos los nodos cuyo valor sea mayor que un número `K`**.

La función debe modificar la lista **sin crear una nueva lista**.

## Entrada

Una lista enlazada simple:

```
4 → 7 → 2 → 9 → 1 → 6
```

y un número:

```
K = 5
```

## Salida esperada

```
4 → 2 → 1
```





## Punto 1 — Eliminar nodos mayores que K

## Lista enlazada simple

## Solución

```python
from typing import Any

def remove_greater_than(linked_list: SinglyLinkedList, k: Any) -> None:
  # 1. Eliminar nodos al inicio mientras sean mayores que k
  while linked_list.head is not None and linked_list.head.data > k:
    linked_list.head = linked_list.head.next

  # 2. Si la lista quedó vacía, también hay que actualizar el tail
  if linked_list.head is None:
    linked_list.tail = None
    return

  # 3. Recorrer desde el primer nodo válido
  current = linked_list.head

  while current.next is not None:
    if current.next.data > k:
      current.next = current.next.next
    else:
      current = current.next

  # 4. Recalcular el tail
  linked_list.tail = current
```

## Explicación paso a paso

La idea aquí es **eliminar nodos sin crear una lista nueva**.

Primero se resuelve el caso más delicado: cuando los nodos a eliminar están al inicio.
Eso obliga a mover `head` hacia adelante mientras el valor del primer nodo sea mayor que `k`.

Luego hay que verificar si la lista quedó vacía. Esto puede pasar, por ejemplo, si todos los nodos eran mayores que `k`. En ese caso, tanto `head` como `tail` deben quedar en `None`.

Después se recorre la lista con un puntero `current`.
No se compara `current.data`, sino `current.next.data`, porque así se puede decidir si el siguiente nodo debe ser saltado:

```python
current.next = current.next.next
```

Eso elimina el nodo intermedio de la cadena.

Finalmente, el `tail` debe apuntar al último nodo real de la lista. Como el recorrido termina con `current` en el último nodo válido, basta con hacer:

```python
linked_list.tail = current
```

## Punto 2 — Detectar el primer valor repetido (Lista Enlazada Simple)

Dada una **lista enlazada simple de números enteros**, determine **cuál es el primer valor que aparece más de una vez en la lista**.

Si ningún valor se repite, la función debe retornar `None`.

## Entrada

Lista enlazada:

```
3 → 7 → 4 → 9 → 7 → 2
```

## Salida

```
7
```

porque es el **primer valor que aparece nuevamente** al recorrer la lista.

## Otro ejemplo

Entrada:

```
1 → 5 → 8 → 3
```

Salida:

```
None
```

## Punto 2 — Detectar el primer valor repetido

## Lista enlazada simple

## Solución

```python
from typing import Optional

def first_repeated_value(linked_list: SinglyLinkedList) -> Optional[Any]:
  seen = set()
  current = linked_list.head

  while current is not None:
    if current.data in seen:
      return current.data
    seen.add(current.data)
    current = current.next

  return None
```

## Explicación paso a paso

Aquí el objetivo no es hallar el valor con mayor frecuencia, sino **el primer valor que reaparece al recorrer la lista de izquierda a derecha**.

Se usa un conjunto `seen` porque permite saber rápidamente si un valor ya apareció antes.

El recorrido funciona así:

* Si el dato actual **ya está** en `seen`, entonces ese es el primer repetido y se retorna.
* Si no está, se agrega al conjunto y se continúa.

Por ejemplo, en:

```python
3 → 7 → 4 → 9 → 7 → 2
```

el recorrido va guardando `3`, `7`, `4`, `9`.
Cuando vuelve a aparecer `7`, se detecta que ya estaba en `seen`, así que se retorna `7`.

La complejidad temporal es buena porque cada consulta al conjunto es, en promedio, O(1).

```
True
```


Salida:

```
False
```

## Punto 3 — Verificar si una lista doblemente enlazada es palíndroma

## Lista enlazada doble

## Solución

```python
def is_palindrome_doubly(linked_list: DoublyLinkedList) -> bool:
  left = linked_list.head
  right = linked_list.tail

  while left is not None and right is not None:
    if left.data != right.data:
      return False

    # Si se cruzan o se encuentran, ya se revisó todo
    if left == right or left.next == right:
      break

    left = left.next
    right = right.prev

  return True
```

## Explicación paso a paso

Este ejercicio aprovecha lo más valioso de una lista doblemente enlazada:
se puede avanzar desde el inicio y retroceder desde el final.

Se usan dos punteros:

* `left`, que arranca en `head`
* `right`, que arranca en `tail`

En cada iteración se comparan sus valores:

* Si son diferentes, la lista no es palíndroma.
* Si son iguales, ambos punteros avanzan hacia el centro.

La condición de parada importante es cuando:

* ambos punteros quedan en el mismo nodo, o
* quedan uno al lado del otro

Eso significa que ya se revisó toda la estructura.

Por ejemplo, en:

```python
1 ⇄ 4 ⇄ 7 ⇄ 4 ⇄ 1
```

se compara:

* `1` con `1`
* `4` con `4`
* se llega al centro y termina

## Casos de prueba
