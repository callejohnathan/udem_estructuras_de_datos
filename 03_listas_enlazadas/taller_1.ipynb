## 🟢 Ejercicio 1: Lista Enlazada Simple

**Enunciado:**
Implementa una función que reciba una **lista enlazada simple** de enteros y un valor `x`. Tu tarea es **eliminar el primer nodo que contenga ese valor**.

* Si el valor no existe, la lista debe permanecer igual.
* No puedes usar listas de Python (`list`) para resolver el problema, solo la implementación de lista enlazada.

**Ejemplo:**

```
Entrada: [5 → 8 → 3 → 10], x = 3  
Salida: [5 → 8 → 10]
```

👉 **Punto de aprendizaje**: práctica con el manejo de punteros, casos especiales (eliminación en cabeza, en medio, o cuando no existe el valor).

---

## 🔵 Ejercicio 2: Lista Doblemente Enlazada

**Enunciado:**
Dada una **lista doblemente enlazada**, implementa una función que **inserte un nodo con valor `x` en la posición `p` (basada en índice 0)**.

* Si `p = 0`, se inserta al inicio.
* Si `p` es mayor al tamaño de la lista, se inserta al final.

**Ejemplo:**

```
Entrada: [1 ⇄ 2 ⇄ 4], x = 3, p = 2  
Salida: [1 ⇄ 2 ⇄ 3 ⇄ 4]
```

👉 **Punto de aprendizaje**: reforzar la diferencia entre `prev` y `next`, y el manejo de enlaces dobles.





---

## 🔴 Ejercicio 3: Lista Circular

**Enunciado:**
Se tiene una **lista circular simple** que representa a jugadores en un juego. Implementa una función que, dado un número `k`, **simule `k` pasos de turno y devuelva el valor del jugador al que le toca jugar**.

**Ejemplo:**

```
Jugadores: [A → B → C → D → (A otra vez)], k = 5  
Salida: "B"  
(orden: A → B → C → D → A → B)
```

👉 **Punto de aprendizaje**: aplicar la circularidad para recorrer indefinidamente, reforzar el concepto de que nunca hay `None`.
