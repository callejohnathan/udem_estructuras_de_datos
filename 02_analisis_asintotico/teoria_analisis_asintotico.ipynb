## 1️⃣ ¿Por qué medir la eficiencia de un algoritmo?

Imagina que escribes dos soluciones distintas para el mismo problema. Ambas dan el resultado correcto... pero una tarda 2 milisegundos y la otra tarda 3 horas con la misma entrada.

> **El análisis de complejidad responde una pregunta muy concreta: ¿qué le pasa al tiempo (o a la memoria) que consume tu algoritmo cuando el tamaño de la entrada `n` crece?**

No nos interesa "cuántos segundos tarda en mi computador" — eso depende del hardware, del lenguaje, de si tienes más procesos corriendo al mismo tiempo. Nos interesa **la tendencia de crecimiento**: si `n` se duplica, ¿el trabajo se duplica también? ¿se cuadruplica? ¿se dispara?

🔎 Frase para quedártela:

> *"Big O no mide velocidad. Mide cómo escala tu solución."*

## 2️⃣ Complejidad temporal vs. complejidad espacial

| | ¿Qué mide? | Ejemplo de lo que cuenta |
|---|---|---|
| ⏱️ **Complejidad temporal** | Cuántas operaciones "básicas" ejecuta el algoritmo en función de `n` | Comparaciones, sumas, llamadas a función, iteraciones |
| 📦 **Complejidad espacial** | Cuánta memoria adicional necesita el algoritmo en función de `n` | Variables nuevas, listas auxiliares, la pila de llamadas recursivas |

Un algoritmo puede ser rápido en tiempo pero costoso en memoria (o viceversa) — normalmente hay que analizar **ambas por separado**, y muchas veces hay un *trade-off* entre ellas.

## 3️⃣ Notación: O, Ω, Θ — ¿y por qué casi siempre usamos O?

Cuando analizamos un algoritmo, en realidad hay tres preguntas distintas:

| Notación | Nombre | Pregunta que responde | Caso |
|---|---|---|---|
| `O(f(n))` | **Big O** (cota superior) | ¿Cuánto es lo *peor* que puede tardar? | Peor caso |
| `Ω(f(n))` | **Omega** (cota inferior) | ¿Cuánto es lo *mínimo* que va a tardar sí o sí? | Mejor caso |
| `Θ(f(n))` | **Theta** (cota ajustada) | ¿Cuál es el comportamiento *exacto*, ni más ni menos? | Cuando peor y mejor caso coinciden |

📌 En la práctica, casi siempre hablamos de **Big O** porque nos interesa **garantizar** qué tan mal se puede poner un algoritmo — es la cota que te protege de sorpresas cuando `n` crece.

**Ejemplo rápido — búsqueda lineal en una lista:**
* Mejor caso (Ω): el elemento buscado está de primero → 1 comparación → `Ω(1)`.
* Peor caso (O): el elemento no está, o está de último → recorres todo → `O(n)`.
* Como el mejor y el peor caso **no coinciden**, no hay un `Θ` limpio para el caso general — por eso decimos que la búsqueda lineal es `O(n)`, usando siempre el peor caso como referencia por defecto.

## 4️⃣ Reglas prácticas para calcular Big O

Cuando cuentas operaciones, hay 3 reglas que simplifican casi cualquier análisis:

1. **Quédate con el término dominante.** Si tu algoritmo hace `n² + n + 1` operaciones, el `n²` es el que manda cuando `n` crece — los demás términos se vuelven insignificantes en comparación. Por eso decimos que es `O(n²)`, no `O(n² + n + 1)`.
2. **Ignora las constantes multiplicativas.** `O(3n)` y `O(n)` describen el mismo crecimiento — la constante no cambia la *forma* de la curva. Por eso `O(3n)` se escribe simplemente `O(n)`.
3. **Ciclos secuenciales se suman, ciclos anidados se multiplican.**
   * Dos ciclos uno *después* del otro (no anidados): `O(n) + O(n) = O(n)` (la suma de dos términos del mismo orden sigue siendo ese orden).
   * Un ciclo *dentro* de otro: `O(n) × O(n) = O(n²)`.

Con estas 3 reglas puedes analizar la gran mayoría del código que vas a escribir en este curso.

## 5️⃣ Clases de complejidad más comunes

| Notación | Nombre | Ejemplo típico |
|---|---|---|
| `O(1)` | Constante | Acceder a `lista[i]`, `diccionario[clave]` |
| `O(log n)` | Logarítmica | Búsqueda binaria |
| `O(n)` | Lineal | Recorrer una lista una vez |
| `O(n log n)` | Cuasi-lineal | Buenos algoritmos de ordenamiento (`sorted()`, mergesort) |
| `O(n²)` | Cuadrática | Dos ciclos anidados sobre la misma entrada |
| `O(2ⁿ)` | Exponencial | Recursión que se ramifica en 2 sin memoization (ej. Fibonacci ingenuo) |
| `O(n!)` | Factorial | Generar todas las permutaciones de `n` elementos |

### 📈 El golpe de realidad: cómo crecen en la práctica

Supongamos que cada operación tarda 1 microsegundo (`µs`):

| `n` | `O(log n)` | `O(n)` | `O(n log n)` | `O(n²)` | `O(2ⁿ)` |
|---|---|---|---|---|---|
| 10 | ~3 µs | 10 µs | ~33 µs | 100 µs | 1 024 µs (~1 ms) |
| 20 | ~4 µs | 20 µs | ~86 µs | 400 µs | ~1 segundo |
| 50 | ~6 µs | 50 µs | ~282 µs | 2 500 µs | ~35 792 años 😱 |

🔴 **Conclusión:** un algoritmo puede ser 100% correcto... y aun así ser completamente inutilizable si su complejidad crece demasiado rápido para el tamaño real de tus datos.

### 🧩 Ejemplo 1 — `O(1)`: acceso directo

**Análisis:** sin importar si `lista` tiene 5 o 5 millones de elementos, acceder a `lista[0]` toma **el mismo tiempo** — es una operación de acceso directo por índice, no un recorrido.

✅ Complejidad temporal: `O(1)`
✅ Complejidad espacial: `O(1)` (no se crea ninguna estructura nueva)

````python
def primer_elemento(lista: list[int]) -> int:
    return lista[0]


print(primer_elemento([4, 8, 15, 16, 23, 42]))
````

### 🧩 Ejemplo 2 — `O(log n)`: búsqueda binaria

**Análisis:** en cada vuelta del `while`, **descartamos la mitad** de lo que quedaba por revisar. Si empiezas con `n` elementos, tras una vuelta quedan `n/2`, luego `n/4`, luego `n/8`... hasta llegar a 1.

¿Cuántas veces puedes dividir `n` entre 2 hasta llegar a 1? Exactamente `log₂(n)` veces.

✅ Complejidad temporal: `O(log n)` — **requiere que la lista esté ordenada**.
✅ Complejidad espacial: `O(1)` (solo variables auxiliares, ninguna estructura crece).

````python
def busqueda_binaria(lista: list[int], objetivo: int) -> int:
    izq, der = 0, len(lista) - 1
    while izq <= der:
        medio = (izq + der) // 2
        if lista[medio] == objetivo:
            return medio
        elif lista[medio] < objetivo:
            izq = medio + 1
        else:
            der = medio - 1
    return -1


print(busqueda_binaria([1, 3, 5, 7, 9, 11, 13], 9))
````

### 🧩 Ejemplo 3 — `O(n)`: recorrido lineal

**Análisis:** el ciclo `for` visita cada elemento de `nums` exactamente una vez. Si `nums` tiene el doble de elementos, el ciclo hace el doble de trabajo — crecimiento **lineal**.

✅ Complejidad temporal: `O(n)`
✅ Complejidad espacial: `O(1)` (una sola variable acumuladora, sin importar `n`)

````python
def suma_total(nums: list[int]) -> int:
    total = 0
    for x in nums:
        total += x
    return total


print(suma_total([1, 2, 3, 4, 5]))
````

### 🧩 Ejemplo 4 — `O(n log n)`: ordenar una lista

**Análisis:** ordenar no es lineal ni cuadrático — los buenos algoritmos de ordenamiento (como el que usa `sorted()` internamente, o *mergesort*) dividen la lista a la mitad repetidamente (`log n` niveles de división) y en cada nivel recorren y combinan `n` elementos. Por eso el resultado es `n` multiplicado por `log n`.

Esta es la misma idea de **dividir y vencerás** que vas a ver formalmente más adelante en el curso.

✅ Complejidad temporal: `O(n log n)`
✅ Complejidad espacial: depende de la implementación — `sorted()` en Python usa `O(n)` espacio adicional (crea una lista nueva).

````python
numeros = [5, 3, 8, 1, 9, 2]
ordenados = sorted(numeros)
print(ordenados)
````

### 🧩 Ejemplo 5 — `O(n²)`: ciclos anidados

**Análisis:** por cada elemento `i`, el ciclo interno vuelve a recorrer **todos** los elementos otra vez. Si `nums` tiene `n` elementos, se hacen `n × n` comparaciones en el peor caso — de ahí el nombre "cuadrático".

✅ Complejidad temporal: `O(n²)`
✅ Complejidad espacial: `O(1)` (el resultado es un booleano, no crece con `n`)

````python
def existe_par_repetido(nums: list[int]) -> bool:
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] == nums[j]:
                return True
    return False


print(existe_par_repetido([1, 5, 3, 5, 2]))
````

### 🧩 Ejemplo 6 — `O(2ⁿ)`: recursión exponencial

**Análisis:** cada llamada a `fibonacci(n)` genera **dos** llamadas nuevas (`fibonacci(n-1)` y `fibonacci(n-2)`), que a su vez generan dos más cada una, y así sucesivamente. El árbol de llamadas se duplica en cada nivel de profundidad.

⚠️ Prueba a correr `fibonacci(35)` en la celda de abajo y compara cuánto tarda contra `fibonacci(20)` — esa diferencia **es** la complejidad exponencial en acción.

✅ Complejidad temporal: `O(2ⁿ)` (sin memoization — más adelante en el curso vas a ver cómo optimizar esto)
✅ Complejidad espacial: `O(n)` — aunque el número de llamadas es exponencial, la **profundidad máxima** de la pila de llamadas activas al mismo tiempo es solo `n`.

````python
def fibonacci(n: int) -> int:
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


print(fibonacci(20))
````

## 7️⃣ Un vistazo más de cerca a la complejidad espacial

Dos funciones pueden resolver "lo mismo" y aun así tener un costo de memoria completamente distinto:

```python
# Opción A: crea una lista nueva
def duplicar_valores(nums: list[int]) -> list[int]:
    resultado = []
    for x in nums:
        resultado.append(x * 2)
    return resultado

# Opción B: modifica la lista original
def duplicar_en_sitio(nums: list[int]) -> None:
    for i in range(len(nums)):
        nums[i] *= 2
```

Ambas son `O(n)` en **tiempo** (recorren `nums` una vez). Pero en **espacio**:

| Función | Complejidad espacial | Por qué |
|---|---|---|
| `duplicar_valores` | `O(n)` | Crea una lista `resultado` nueva del mismo tamaño que `nums` |
| `duplicar_en_sitio` | `O(1)` | Modifica `nums` directamente, sin ninguna estructura auxiliar que crezca |

💡 Esta diferencia importa mucho cuando `nums` es enorme (piensa en millones de registros) — duplicar la memoria usada puede ser la diferencia entre que tu programa corra o se quede sin RAM.
