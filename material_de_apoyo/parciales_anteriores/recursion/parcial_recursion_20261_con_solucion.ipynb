# Punto 1 — Suma en listas anidadas (profundidad variable)

### Enunciado

Dada una estructura de datos que es una **lista de listas de listas...** (profundidad variable) que contiene **enteros positivos**, calcula la **sumatoria de todos los números estrictamente mayores a K**.

La estructura puede tener sublistas con diferentes profundidades (por ejemplo, una lista puede contener números directamente y también listas anidadas).

### Entrada

1. Una estructura de listas anidadas (representada como literal estilo Python, por ejemplo: `[1, [2, [3]]]`)
2. Un entero positivo `K`

### Salida

Un entero: la suma de todos los valores `> K` encontrados en toda la estructura.

### Ejemplo 1

**Entrada:**

```
L = [1, [2, [3, 10], 4], [[5], 9], 8]
K = 4
```

**Salida:**

```
32
```

**Explicación breve:** se suman `10 + 5 + 9 + 8 = 32` (todos son > 4).

## ✅ Solución

```python
from __future__ import annotations
from typing import Union, List

NestedIntList = List[Union[int, "NestedIntList"]]


def sum_greater_than_k(L: NestedIntList, k: int) -> int:
    """
    Retorna la suma de todos los enteros estrictamente mayores a k
    dentro de una lista anidada de profundidad variable.
    Restricciones: sin ciclos, sin slicing.
    """
    return _sum_list(L, k, 0)


def _sum_list(L: NestedIntList, k: int, i: int) -> int:
    """
    Recorre recursivamente la lista L usando un índice i.
    i representa "en qué posición voy", para evitar ciclos y slicing.
    """
    # Caso base: si ya llegamos al final de esta lista, no hay más que sumar.
    if i == len(L):
        return 0

    current = L[i]

    # Si es un entero, decidimos si suma o no según k.
    if isinstance(current, int):
        add = current if current > k else 0
        # Importante: avanzamos al siguiente elemento aumentando i en 1.
        return add + _sum_list(L, k, i + 1)

    # Si no es int, asumimos que es una sublista:
    # sumamos lo que aporte esa sublista + lo que aporte el resto de la lista actual.
    return _sum_list(current, k, 0) + _sum_list(L, k, i + 1)
```

```python
K = 4

```

## 🧠 Qué es lo “crítico” de esta solución (por qué está bien para el parcial)

* **Sin ciclos:** el “recorrido” se hace con `_sum_list(L, k, i)` avanzando `i + 1`.
* **Sin slicing:** no hacemos `L[1:]`; usamos `i` para ir elemento por elemento.
* **Profundidad variable:** si `current` es lista, llamamos `_sum_list(current, k, 0)` y entramos a ese nivel.
* **Correctitud:** cada elemento se procesa **exactamente una vez** en su lista correspondiente.

# Punto 2 — Máximo por fila en matriz n×m con recursión de cola

### Enunciado

Dada una matriz de tamaño **n×m** (número de filas diferente al de columnas), devuelve una lista con el **valor máximo de cada fila**.

Restricciones obligatorias:

* **No puedes usar** `max`, `sort` ni funciones equivalentes para obtener máximos.
* Debes implementar la solución usando **recursión de cola (tail recursion)**.

### Entrada

* Una matriz n*m de enteros.

### Salida

Una lista con `n` valores: el máximo de cada fila.

### Ejemplo 1

**Entrada:**

```
M = [[1, 9, 3, 4],[-2, -8, -1, -5],[10, 10, 7, 8]]
```

**Salida:**

```
[9, -1, 10]
```



---

La estrategia es 100% tail-recursive en dos niveles:

1. `_max_in_row_tail(row, j, current_max)` recorre una fila por índice y va guardando el máximo parcial.
2. `_max_each_row_tail(M, i, acc)` recorre las filas y va acumulando los máximos en `acc`.

> Nota: para *agregar* al acumulador sin usar `append` (que muta) podemos hacer `acc + [value]` (esto NO es slicing). Si te dejan mutar, también es válido usar `append`, pero aquí lo dejo “funcional” y claro.

---

## ✅ Solución (tail recursion, sin ciclos/slicing, sin max/sort)

```python
from __future__ import annotations
from typing import List


def max_per_row(M: List[List[int]]) -> List[int]:
    """
    Devuelve una lista con el máximo de cada fila de M.
    Restricciones: sin ciclos, sin slicing, sin max/sort.
    Implementación: recursión de cola (acumuladores).
    """
    return _max_each_row_tail(M, i=0, acc=[])


def _max_each_row_tail(M: List[List[int]], i: int, acc: List[int]) -> List[int]:
    """
    Tail recursion sobre las filas:
    - i: índice de la fila actual
    - acc: acumulador con los máximos ya calculados
    """
    # Caso base: ya procesamos todas las filas
    if i == len(M):
        return acc

    row = M[i]
    # Asumimos filas no vacías; si pudieran venir vacías, habría que definir regla.
    row_max = _max_in_row_tail(row, j=1, current_max=row[0])

    # Tail call: el "trabajo" ya quedó hecho en los parámetros (nuevo i y nuevo acc)
    return _max_each_row_tail(M, i + 1, acc + [row_max])


def _max_in_row_tail(row: List[int], j: int, current_max: int) -> int:
    """
    Tail recursion sobre una fila:
    - j: índice del elemento a comparar
    - current_max: máximo parcial encontrado hasta ahora
    """
    # Caso base: llegamos al final de la fila
    if j == len(row):
        return current_max

    value = row[j]

    # Línea clave: reemplazamos el máximo parcial sin usar max()
    new_max = value if value > current_max else current_max

    # Tail call: avanzamos j, y llevamos el nuevo máximo parcial
    return _max_in_row_tail(row, j + 1, new_max)
```

---

## 🧪 Prueba con tu ejemplo

```python
M = [
    [1, 9, 3, 4],
    [-2, -8, -1, -5],
    [10, 10, 7, 8]
]
print(max_per_row(M))  # [9, -1, 10]
```

---

## 🧠 Por qué esto sí es recursión de cola (tail recursion)

* En `_max_in_row_tail(...)`, la **última operación** es el llamado recursivo; no queda nada “pendiente” por hacer luego.
* En `_max_each_row_tail(...)` igual: calculo `row_max`, lo meto en el acumulador y el último paso es la llamada recursiva.

### Enunciado

Dado un string compuesto por **letras y números**, elimina del string **los números pares** (0, 2, 4, 6, 8) **solo si** cumplen esta condición:

✅ El dígito par debe tener **una letra inmediatamente a la izquierda y una letra inmediatamente a la derecha**.

Restricciones obligatorias:

* **No puedes convertir tipos**, por ejemplo: **no puedes usar `int()`**.


Notas:

* Si el dígito par está al inicio o al final, **no se elimina** (no tiene letras a ambos lados).
* Si al lado hay otro número (en vez de letra), **no se elimina**.

### Entrada

Un string `S`

### Salida

El string resultante luego de eliminar los dígitos pares que estén “encerrados” entre letras.

### Ejemplo 1

**Entrada:**

```
a2b
```

**Salida:**

```
ab
```

### Ejemplo 2

**Entrada:**

```
ab4c7d
```

**Salida:**

```
abc7d
```

(Se elimina `4` porque está entre `b` y `c`; `7` no se elimina porque es impar.)


---




La idea es **recorrer el string por índice** y decidir si cada carácter se mantiene o se elimina.

Para eliminar un dígito necesitamos verificar tres cosas:

1. Que sea un **dígito par** (`0,2,4,6,8`)
2. Que **no esté en los extremos**
3. Que **los vecinos sean letras**

---

# ✅ Solución Recursiva

```python
from typing import Final


EVEN_DIGITS: Final[str] = "02468"


def remove_even_between_letters(s: str) -> str:
    """
    Elimina los dígitos pares que estén encerrados entre letras.
    Restricciones: sin ciclos, sin slicing, sin conversiones de tipo.
    """
    return _process(s, 0)


def _process(s: str, i: int) -> str:
    """
    Recorre el string recursivamente usando un índice.
    """
    
    # Caso base: llegamos al final del string
    if i == len(s):
        return ""

    current = s[i]

    # Verificamos si el carácter actual es un dígito par
    if _is_even_digit(current) and _is_between_letters(s, i):
        # Si cumple la condición de eliminación, simplemente
        # NO lo agregamos al resultado y seguimos procesando
        return _process(s, i + 1)

    # En cualquier otro caso mantenemos el carácter
    return current + _process(s, i + 1)


def _is_even_digit(c: str) -> bool:
    """
    Determina si un carácter es un dígito par sin convertir a int.
    """
    return c in EVEN_DIGITS


def _is_between_letters(s: str, i: int) -> bool:
    """
    Verifica si el carácter en posición i está entre dos letras.
    """
    
    # Si está en los extremos, automáticamente no se elimina
    if i == 0 or i == len(s) - 1:
        return False

    left = s[i - 1]
    right = s[i + 1]

    # Deben ser letras a ambos lados
    return left.isalpha() and right.isalpha()
```

# 🧪 Pruebas

```python
# ab

print(remove_even_between_letters("ab4c7d"))
# abc7d

# 2abc

print(remove_even_between_letters("a8b2c"))
# abc
```

# 🧠 Idea Recursiva Clave

El patrón es:

```
resultado = caracter_actual + resultado_del_resto
```

pero **si el carácter se elimina**, entonces:

```
resultado = resultado_del_resto
```

Esto produce una estructura conceptual como:

```
_process("a2b",0)
  "a" + _process("a2b",1)
            _process("a2b",1) elimina '2'
            -> _process("a2b",2)
                     "b" + _process("a2b",3)
```

Resultado final:

```
ab
```

````python
def delete_(string: str, pos: int = 1, current_string: str = "") -> str:
  if(len(string) <= 1):
    return string

  if(pos == len(string)-1):
    return string[0] + current_string + string[pos]

  char = string[pos]
  if(char.isdigit() and char in "02468"):
    if(string[pos-1].isalpha() and string[pos+1].isalpha()):
      return delete_(string, pos+1, current_string)

  return delete_(string, pos+1, current_string + char)

delete_("") #ab
````
