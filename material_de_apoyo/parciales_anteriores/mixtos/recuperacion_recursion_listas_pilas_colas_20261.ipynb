## Punto 1 — Recursión (50 pts)

**Contexto:** En un sistema de análisis estadístico, se procesan matrices cuadradas de datos numéricos. Para resumir la información de cada fila, se desea calcular la sumatoria parcial de sus primeros elementos usando un enfoque recursivo puro.

**Enunciado:**

Implemente una función **recursiva sin cola** que, dada una matriz cuadrada de enteros de tamaño `n×n` y un entero `k` (con `1 ≤ k ≤ n`), calcule y retorne la **sumatoria de los primeros `k` elementos de cada fila**.

La función debe retornar una lista (o arreglo) de `n` enteros, donde la posición `i` contiene la suma de los primeros `k` elementos de la fila `i`.

**Restricciones:**
- La solución debe ser recursiva y **no debe usar recursión de cola**: el resultado de la llamada recursiva debe participar en un cómputo al momento de retornar (no simplemente devolverse directamente).
- No se permite el uso de ciclos (`for`, `while`) dentro de la función recursiva.
- Puede usar funciones auxiliares recursivas si lo considera necesario (por ejemplo, una para recorrer columnas y otra para recorrer filas), pero **ninguna de ellas puede ser de cola**.

---

**Ejemplo de prueba:**

Matriz `4×4`:

```
 2   5   1   8
 3   7   4   6
 9   2   3   5
 1   4   8   2
```

`k = 3`

**Procesamiento fila por fila:**

| Fila | Primeros k=3 elementos | Sumatoria |
|------|------------------------|-----------|
| 0 | 2, 5, 1 | **8** |
| 1 | 3, 7, 4 | **14** |
| 2 | 9, 2, 3 | **14** |
| 3 | 1, 4, 8 | **13** |

**Resultado esperado:**

```
[8, 14, 14, 13]
```


## Punto 2 — Listas Enlazadas Dobles (50 pts)

**Contexto:** Un sistema de moderación de comentarios almacena los mensajes de los usuarios en una lista doblemente enlazada. Se necesita eliminar las primeras ocurrencias de mensajes que contengan cierta palabra clave.

**Enunciado:**

Dada una **lista enlazada doble** de Strings, un String `s` y un entero `k`, elimine los **primeros `k` nodos** cuyo valor **contenga** a `s` como subcadena (no necesariamente igual, basta con que `s` esté incluido en el valor del nodo).

Una vez se hayan eliminado `k` nodos, detenga el recorrido aunque queden nodos con `s` más adelante. Si hay menos de `k` nodos que contengan a `s`, elimine todos los que encuentre.

**Restricciones:**
- Debe aprovechar los punteros `next` y `prev` propios de la lista doble para hacer las eliminaciones directamente, sin copiar la lista a otra estructura.
- La comparación no distingue mayúsculas de minúsculas (por ejemplo, `"spam"` debe coincidir con `"SPAM_message"` y con `"AntiSpam"`).
- Retorne la lista modificada.

---

**Ejemplo de prueba:**

Lista inicial:

```
"hola mundo" ↔ "spam detectado" ↔ "mensaje limpio" ↔ "SPAM bloqueado" ↔ "otro spam aquí" ↔ "todo bien"
```

`s = "spam"`, `k = 2`

**Procesamiento:**

| Nodo visitado | ¿Contiene "spam"? | Acción | Eliminaciones restantes |
|---|---|---|---|
| `"hola mundo"` | ❌ | Se omite | 2 |
| `"spam detectado"` | ✅ | Se elimina | 1 |
| `"mensaje limpio"` | ❌ | Se omite | 1 |
| `"SPAM bloqueado"` | ✅ | Se elimina | 0 → Se detiene |

**Lista resultante:**

```
"hola mundo" ↔ "mensaje limpio" ↔ "otro spam aquí" ↔ "todo bien"
```

> Nota: `"otro spam aquí"` **no se elimina** porque ya se alcanzó el límite `k = 2`.

## Punto 3 — Pilas (50 pts)

**Contexto:** En un sistema de validación de transacciones financieras, se tiene un historial de montos almacenados en una pila. Se desea saber si existe algún par de montos cuya suma o diferencia sea igual a un valor objetivo.

**Enunciado:**

Dada una **pila de enteros** `p` y un entero `k`, determine si existe algún par de elementos distintos `(a, b)` dentro de la pila tal que:

- `a + b = k`, **o**
- `a - b = k`

Si existe al menos un par que cumpla alguna de las dos condiciones, retorne una estructura con los dos valores y la operación encontrada (`"suma"` o `"resta"`). Si se cumplen ambas condiciones para el mismo par u otros pares, retorne cualquiera de las respuestas válidas. Si no existe ningún par que cumpla ninguna condición, retorne `-1`.

**Restricciones:**
- Solo puede usar operaciones propias de pila (`push`, `pop`, `peek`) y pilas auxiliares. No se permite copiar el contenido a arreglos, listas u otras estructuras.
- Los dos elementos del par deben ser nodos distintos en posición (aunque pueden tener el mismo valor).
- La pila `p` debe quedar en su **estado original** al finalizar.

---

**Ejemplo de prueba:**

```
p (tope → fondo):  10 | 3 | 7 | 5 | 2
k = 8
```

**Pares posibles a verificar:**

| Par (a, b) | a + b | a - b |
|---|---|---|
| (10, 3) | 13 | 7 |
| (10, 7) | 17 | 3 |
| (10, 5) | 15 | **5** → no es 8 |
| (10, 2) | 12 | 8 ✅ |
| (3, 7) | 10 | -4 |
| (3, 5) | **8** ✅ | -2 |
| ... | ... | ... |

Se encuentran dos pares válidos. El algoritmo puede retornar cualquiera de:

```
Resultado: (10, 2) → resta → 10 - 2 = 8
```
o
```
Resultado: (3, 5) → suma → 3 + 5 = 8
```

**Caso sin solución:** con `k = 100` → Retorna `-1`.

## Punto 4 — Colas (50 pts)

**Contexto:** En la sala de espera de un centro médico, las personas hacen fila para ser atendidas. Por política del centro, si una persona es mayor que **todas** las que tiene delante en la fila, se le pide que pase directamente (se retira de la cola), pues se considera que ya ha esperado suficiente proporcionalmente a su edad.

**Enunciado:**

Dada una **cola de personas**, donde cada persona tiene al menos los atributos `nombre` (String) y `edad` (int), elimine de la cola a toda persona que sea **estrictamente mayor** que todas las personas que tiene delante de ella en la cola (es decir, todas las personas que llegaron antes).

La persona al frente de la cola nunca se elimina, ya que no tiene antecesores.

Una vez procesada la cola, retorne la cola resultante con las personas que permanecen.

**Restricciones:**
- Solo puede usar operaciones propias de cola (`enqueue`, `dequeue`, `peek`) y colas o pilas auxiliares. No se permite copiar el contenido a arreglos o listas.
- El orden relativo de las personas que permanecen debe conservarse.
- Se considera "antecesores" a todas las personas que están delante en la cola al momento de procesar, **no solo la inmediatamente anterior**.

---

**Ejemplo de prueba:**

Cola inicial (frente → fondo):

```
Ana(32) → Luis(45) → Marta(28) → Jorge(51) → Sofía(60) → Pedro(40)
```

**Análisis por persona:**

| Persona | Edad | Antecesores (al frente) | ¿Mayor que TODOS? | Acción |
|---|---|---|---|---|
| Ana (32) | 32 | Ninguno | — | ✅ Permanece |
| Luis (45) | 45 | Ana(32) | 45 > 32 ✅ | ❌ Se elimina |
| Marta (28) | 28 | Ana(32) | 28 > 32 ❌ | ✅ Permanece |
| Jorge (51) | 51 | Ana(32), Marta(28) | 51 > 32 y 51 > 28 ✅ | ❌ Se elimina |
| Sofía (60) | 60 | Ana(32), Marta(28) | 60 > 32 y 60 > 28 ✅ | ❌ Se elimina |
| Pedro (40) | 40 | Ana(32), Marta(28) | 40 > 32 pero 40 > 28 ✅ | ❌ Se elimina |

> ⚠️ Pedro(40): 40 > 32 ✅ y 40 > 28 ✅ → mayor que **todos** sus antecesores → se elimina.

**Cola resultante:**

```
Ana(32) → Marta(28)
```
