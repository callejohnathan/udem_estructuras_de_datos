## Lista Enlazada Simple

````python
from typing import Optional, Any

class NodeSingly:
  def __init__(self, value: Any, next = None) -> None:
    self.value = value
    self.next = next

  def __repr__(self) -> str:
    return f"{self.value}"

class SinglyLinkedList:
  def __init__(self) -> None:
    self.head: Optional[NodeSingly] = None
    self.tail: Optional[NodeSingly] = None
    self.size: int = 0

  def append(self, value: Any) -> None:
    new_node = NodeSingly(value)
    if not self.head:
      self.head = self.tail = new_node
    else:
      assert self.tail is not None
      self.tail.next = new_node
      self.tail = new_node

    self.size += 1

  def delete_and_return_last(self) -> Any:
    #llegando al penúltimo
    if(self.size == 1):
      old_tail = self.head
      self.head = None
      self.tail = None
      self.size -= 1
      return old_tail.value

    current = self.head
    while(current.next != self.tail):
      current = current.next

    old_tail = self.tail
    self.tail = current
    self.tail.next = None

    self.size -= 1
    return old_tail.value

  def delete_and_return_first(self) -> Any:
    if(self.head is None):
      return None

    old_head = self.head
    self.head = self.head.next
    old_head.next = None
    self.size -= 1
    return old_head.value

  def __repr__(self) -> str:
    values = []
    current = self.head
    while current:
      values.append(str(current.value))
      current = current.next
    return " → ".join(values) if values else "[]"

  def __len__(self) -> int:
    return self.size
````

## Lista Doblemente Enlazada

````python
from typing import Optional, Any

class NodeDoubly:
  def __init__(self, value: Any, prev = None, next = None) -> None:
    self.value = value
    self.prev = prev
    self.next = next

  def __repr__(self) -> str:
    return f"{self.value}"

class DoublyLinkedList:
  def __init__(self) -> None:
    self.head: Optional[NodeDoubly] = None
    self.tail: Optional[NodeDoubly] = None

  def append(self, value: Any) -> None:
    new_node = NodeDoubly(value)
    if not self.head:
      self.head = self.tail = new_node
    else:
      assert self.tail is not None
      self.tail.next = new_node
      new_node.prev = self.tail
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    current = self.head
    while current:
      values.append(str(current.value))
      current = current.next
    return " ⇄ ".join(values) if values else "[]"
````

## Pila implementada con Lista Enlazada

````python
from typing import Any
class Stack:
  def __init__(self):
    self.items = SinglyLinkedList() #estructura de datos implementadora

  def push(self, element: Any) -> None: #agrega después del tope (al final)
    self.items.append(element)
    return

  def pop(self) -> Any:
    if(len(self.items) == 0):
      raise Exception("Stack is Empty...")
    return self.items.delete_and_return_last()

  def peek(self):
    if(len(self.items) == 0):
      raise Exception("Stack is Empty...")

    return self.items.tail.value

  def __repr__(self) -> str:
    return str(self.items)
````

## Cola implementada con Lista Enlazada

````python
from typing import Any

class Queue:
  def __init__(self):
    self.__queue = SinglyLinkedList()

  def enqueue(self, element: Any) -> None: #Agrega después del First
    self.__queue.append(element)

  #FIFO -> First in First out
  def dequeue(self):
    if(len(self.__queue) == 0):
      raise Exception("Cola vacía... No podés hacer dequeue")

    return self.__queue.delete_and_return_first()

  def peek(self) -> Any:
    if(len(self.__queue) == 0):
      raise Exception("Cola vacía... No podés hacer peek")

    return self.__queue.head.value

  def __len__(self) -> int:
    return len(self.__queue)

  def __repr__(self) -> str:
    return str(self.__queue)
````

## Cola de Prioridad

````python
from typing import List

class EmptyQueue(Exception):
  ...

#ADT: Abstract Data Types --> Contrato que define una colección
class PriorityQueue:
  def __init__(self, priority): #priority define si es min o max
    self.__queue: List[int] = [] #Estructura de Datos implementadora
    self.__priority: str = priority

  def enqueue(self, element: int) -> None: #Agrega después del First
    if(self.__priority == "min"):
      self.__queue.append(element)
      self.__queue.sort()
    elif(self.__priority == "max"):
      self.__queue.append(element)
      self.__queue.sort(reverse = True)
    else:
      raise ValueError("Grave... Este tipo de cola no existe...")

  def dequeue(self) -> int : #Retorna y elimina el First
    if(len(self.__queue) == 0):
      raise EmptyQueue("Cola vacía... No podés hacer dequeue")

    return self.__queue.pop(0)


  def peek(self) -> int: #Retorna el First
    if(len(self.__queue) == 0):
      raise EmptyQueue("Cola vacía... No podés hacer dequeue")

    return self.__queue[0]

  def __repr__(self) -> str:
    return str(self.__queue)
````

## Punto 1 — Listas Enlazadas

**Contexto:** En un sistema de análisis de señales, se almacenan lecturas de sensores como una lista enlazada simple de enteros. Se sabe que ciertas secuencias de lecturas relevantes están delimitadas por el mismo valor al inicio y al final (como marcadores). Su tarea es procesar esas secuencias.

**Enunciado:**

Dada una lista enlazada simple de enteros, se define un **grupo** como una subsecuencia contigua de nodos cuyo primer y último elemento tienen el mismo valor. El grupo debe tener **al menos 3 nodos** (de lo contrario no hay nodo intermedio que eliminar).

Recorra la lista de izquierda a derecha buscando grupos. Cuando encuentre un grupo válido:

1. Determine la cantidad de nodos que lo conforman.
2. Identifique el **nodo de la mitad**: si el grupo tiene `n` nodos, el nodo del medio es el de la posición `⌊n/2⌋ + 1` (contando desde 1). Si `n` es par, use la mitad inferior: posición `n/2`.
3. Elimine ese nodo de la lista.
4. Continue el recorrido **desde el nodo inmediatamente siguiente al final del grupo** — no se reinicia la búsqueda desde el principio ni se solapan grupos.

**Restricciones:**
- Un nodo puede ser el inicio de un grupo solo si aún no pertenece a ningún grupo ya procesado.
- La lista original se modifica en sitio (no se crea una lista nueva).

---

**Ejemplo de prueba:**

Lista inicial:

```
3 → 7 → 1 → 5 → 3 → 9 → 2 → 6 → 4 → 2 → 8 → 5
```

**Análisis paso a paso:**

- Se parte desde el nodo `3` (índice 0).
  - Buscando hacia adelante un nodo con valor `3`: se encuentra en índice 4.
  - Grupo: `[3, 7, 1, 5, 3]` → 5 nodos.
  - Nodo del medio: posición `⌊5/2⌋ + 1 = 3` → valor `1`.
  - Se elimina el nodo con valor `1`.
  - Se continúa desde el nodo `9` (siguiente al final del grupo).

- Se parte desde `9` (índice 5 original).
  - No hay ningún nodo con valor `9` más adelante → no hay grupo.
  - Se avanza al `2` (índice 6).

- Se parte desde `2` (índice 6).
  - Buscando hacia adelante un nodo con valor `2`: se encuentra en índice 9.
  - Grupo: `[2, 6, 4, 2]` → 4 nodos.
  - Nodo del medio: posición `4/2 = 2` → valor `6`.
  - Se elimina el nodo con valor `6`.
  - Se continúa desde `8` (siguiente al final del grupo).

- `8` y `5`: no forman grupos. Fin del recorrido.

**Lista resultante:**

```
3 → 7 → 5 → 3 → 9 → 2 → 4 → 2 → 8 → 5
```

## Punto 2 — Pilas

**Contexto:** En un sistema de gestión de inventario, los productos son identificados por códigos de texto. Se tiene una pila de inventario `p1` con los productos disponibles, y una pila de solicitudes `p2` con los productos que se desean priorizar. Su tarea es reorganizar `p1` según las solicitudes de `p2`.

**Enunciado:**

Dadas dos pilas de Strings `p1` y `p2`, procese cada elemento de `p2` **en el orden en que aparecen** (de tope a fondo) de la siguiente manera:

- Tome el tope de `p2` y llámelo `S`.
- Busque `S` en `p1`:
  - **Si `S` está en `p1`**: extráigalo de su posición actual y empújelo al **tope de `p1`**.
  - **Si `S` no está en `p1`**: empuje el string `"S no encontrado"` al tope de `p1` (reemplazando `S` por el valor real del string buscado).
- Repita hasta vaciar `p2`.

**Restricciones:**
- Solo se puede operar con operaciones propias de pila: `push`, `pop` y `peek`. Está permitido usar pilas auxiliares.
- El orden relativo de los elementos de `p2` debe respetarse: el primero en salir de `p2` (su tope) es el primero en ser procesado, y por tanto termina más profundo en `p1` que los procesados después.
- Si `S` aparece más de una vez en `p1`, mueva únicamente la **primera ocurrencia** encontrada desde el tope.

---

**Ejemplo de prueba:**

Estado inicial:

```
p1 (tope → fondo):  "tornillo" | "llave" | "tuerca" | "clavo" | "perno"
p2 (tope → fondo):  "clavo" | "arandela" | "tuerca"
```

**Procesamiento:**

| Paso | S tomado de p2 | ¿Está en p1? | Acción sobre p1 |
|------|----------------|--------------|-----------------|
| 1 | `"clavo"` | ✅ Sí | Se extrae `"clavo"` y se lleva al tope |
| 2 | `"arandela"` | ❌ No | Se empuja `"arandela no encontrado"` al tope |
| 3 | `"tuerca"` | ✅ Sí | Se extrae `"tuerca"` y se lleva al tope |

**Estado final de `p1` (tope → fondo):**

```
"tuerca" | "arandela no encontrado" | "clavo" | "tornillo" | "llave" | "perno"
```

`p2` queda vacía.




## Punto 3 — Colas de Prioridad (34 pts)

```
eficiencia = kmRecorridos / galonesConsumidos
```

Un mayor valor de eficiencia indica un mejor rendimiento (más km por galón).

**Restricciones:**
- Debe usar explícitamente una cola de prioridad para la solución; no se aceptan ordenamientos directos de la lista.
- Si dos buses tienen la misma eficiencia, desempate por orden alfabético ascendente de placa.
- Suponga que siempre `K ≤` tamaño de la flota y que `galonesConsumidos > 0`.

---

**Ejemplo de prueba:**

Flota de buses:

`K = 3`
