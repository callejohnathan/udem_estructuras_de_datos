### Grafo No Ponderado - Matriz de Adyacencia (GraphM)

````python
class GraphM:
    def __init__(self):
        self.adj_matrix = [] #relaciones
        self.nodes = [] #etiquetas

    def add_vertex(self, label):
        if label not in self.nodes:
            self.nodes.append(label)

            #ajustar la matriz
            for row in self.adj_matrix:
                row.append(0)
            self.adj_matrix.append([0] * len(self.nodes))

    def add_edge(self, vertex_1, vertex_2, directed = True):
        if(vertex_1 not in self.nodes):
            self.add_vertex(vertex_1)

        if(vertex_2 not in self.nodes):
            self.add_vertex(vertex_2)

        #estoy completamente seguro que vertex1 y vertex2
        pos_v1 = self.nodes.index(vertex_1)
        pos_v2 = self.nodes.index(vertex_2)

        #ya tengo las dos posiciones->agrego la relación
        self.adj_matrix[pos_v1][pos_v2] = 1
        if(not directed):
            self.adj_matrix[pos_v2][pos_v1] = 1

    def __repr__(self):
        result = "   " + " ".join(self.nodes) + "\n"

        for i, row in enumerate(self.adj_matrix):
            result += self.nodes[i] + "  " + " ".join(map(str, row)) + "\n"

        return result
````

### Grafo No Ponderado - Lista de Adyacencia (GraphL)

````python
from typing import Any, Dict, List
from collections import deque

class GraphL:
  def __init__(self):
    self.adj_list: Dict[Any, List[Any]] = {}
    self.size: int = 0

  def add_vertex(self, value: Any) -> None:
    if value in self.adj_list:
      return None  # ya está el nodo
    self.adj_list[value] = []
    self.size += 1

  def add_edge(self, vertex_1: Any, vertex_2: Any, directed: bool = True):
    if vertex_1 not in self.adj_list:
      self.add_vertex(vertex_1)
    if vertex_2 not in self.adj_list:
      self.add_vertex(vertex_2)

    # agregar la arista
    if vertex_2 not in self.adj_list[vertex_1]:
      self.adj_list[vertex_1].append(vertex_2)

    # si no es dirigido, agregar también la inversa
    if not directed and vertex_1 not in self.adj_list[vertex_2]:
      self.adj_list[vertex_2].append(vertex_1)

  def __repr__(self):
    result = ""
    for vertex, neighbors in self.adj_list.items():
      result += f"{vertex}: {neighbors}\n"
    return result
````

### Grafo Ponderado - Lista de Adyacencia (GraphWL)

````python
from typing import Any, Dict, List
from collections import deque

class GraphWL:
  def __init__(self):
    self.adj_list: Dict[Any, List[Any]] = {}
    self.size: int = 0

  def add_vertex(self, value: Any) -> None:
    if value in self.adj_list:
      return None  # ya está el nodo
    self.adj_list[value] = []
    self.size += 1

  def add_edge(self, vertex_1: Any, vertex_2: Any, directed: bool = True, weight: Any = None):
    if vertex_1 not in self.adj_list:
      self.add_vertex(vertex_1)
    if vertex_2 not in self.adj_list:
      self.add_vertex(vertex_2)

    # agregar la arista
    if vertex_2 not in self.adj_list[vertex_1]:
      self.adj_list[vertex_1].append((vertex_2, weight))

    # si no es dirigido, agregar también la inversa
    if not directed and vertex_1 not in self.adj_list[vertex_2]:
      self.adj_list[vertex_2].append((vertex_1, weight))

  def __repr__(self):
    result = ""
    for vertex, neighbors in self.adj_list.items():
      result += f"{vertex}: {neighbors}\n"
    return result
````

## Ejercicio 1 - Rutas de inspección regulada
---



Una autoridad de tránsito ferroviario supervisa una red de líneas de carga modelada como un grafo dirigido ponderado, donde cada nodo representa una estación y cada arista dirigida representa un tramo con un tonelaje máximo permitido como peso. El departamento de operaciones necesita conocer todas las rutas posibles que parten desde una estación de origen `S`, transitan **exclusivamente** por tramos con tonelaje múltiplo de `N`, y recorren **más de `K` estaciones** (incluyendo la de origen). Esto permite planificar inspecciones exhaustivas que cubran tramos suficientemente largos para ser representativos.
