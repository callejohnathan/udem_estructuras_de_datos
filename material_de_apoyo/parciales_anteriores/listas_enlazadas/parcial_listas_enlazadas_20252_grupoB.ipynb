````python
# --------------------- Simplemente Enlazada
from typing import Optional, Any

class NodeSingly:
  def __init__(self, value: Any, next = None) -> None:
    self.value = value
    self.next = next

  def __repr__(self) -> str:
    return f"{self.value}"

class SinglyLinkedList:
  def __init__(self) -> None:
    self.head = None
    self.tail = None

  def append(self, value: Any) -> None:
    new_node = NodeSingly(value)
    if not self.head:
      self.head = self.tail = new_node
    else:
      assert self.tail is not None
      self.tail.next = new_node
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    current = self.head
    while current:
      values.append(str(current.value))
      current = current.next
    return " → ".join(values) if values else "[]"

# --------------------- Doblemente Enlazada
class NodeDoubly:
  def __init__(self, value: Any, prev = None, next = None) -> None:
    self.value = value
    self.prev = prev
    self.next = next

  def __repr__(self) -> str:
    return f"{self.value}"

class DoublyLinkedList:
  def __init__(self) -> None:
    self.head = None
    self.tail = None

  def append(self, value: Any) -> None:
    new_node = NodeDoubly(value)
    if not self.head:
      self.head = self.tail = new_node
    else:
      assert self.tail is not None
      self.tail.next = new_node
      new_node.prev = self.tail
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    current = self.head
    while current:
      values.append(str(current.value))
      current = current.next
    return " ⇄ ".join(values) if values else "[]"


# --------------------- Circular
class NodeCircular:
  def __init__(self, value: Any, next = None) -> None:
    self.value = value
    self.next = next

  def __repr__(self) -> str:
      return f"{self.value}"

class CircularLinkedList:
  def __init__(self) -> None:
    self.head = None
    self.tail = None

  def append(self, value: Any) -> None:
    new_node = NodeCircular(value)
    if not self.head:
      self.head = self.tail = new_node
      new_node.next = new_node  # apunta a sí mismo
    else:
      assert self.tail is not None
      new_node.next = self.head
      self.tail.next = new_node
      self.tail = new_node

  def __repr__(self) -> str:
    values = []
    if not self.head:
      return "[]"
    current = self.head
    while True:
      values.append(str(current.value))
      current = current.next
      if current == self.head:
        break
    return " → ".join(values) + " → (head)"
````

#### 🔧 ¿Cómo se interpreta “n desde el final”?

* Si `n = 1`, se elimina el **primer nodo par que se encuentre desde el final**.
* Si `n = 2`, se elimina el **segundo nodo par que se encuentre desde el final**.
* Y así sucesivamente, hasta el `n`-ésimo par desde el final.

---

```python

```
